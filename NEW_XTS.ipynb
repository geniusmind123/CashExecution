```python
# Extract holdings with required details
holdings = response['result']['RMSHoldings']['Holdings']
output = []

for isin, details in holdings.items():
    output.append({
        'ISIN': isin,
        'RMSHoldingId': details.get('RMSHoldingId'),
        'ClientId': details.get('ClientId'),
        'HoldingQuantity': details.get('HoldingQuantity'),
        'BuyAvgPrice': details.get('BuyAvgPrice'),
        'ExchangeNSEInstrumentId': details.get('ExchangeNSEInstrumentId'),
        'ExchangeBSEInstrumentId': details.get('ExchangeBSEInstrumentId'),
        'ExchangeMSEInstrumentId': details.get('ExchangeMSEInstrumentId'),
    })

# Print output
for item in output:
    print(item)
```

```python
# Function to get holding quantity by instrument ID (either from NSE, BSE, or MSE)
def get_holding_quantity(instrumentid):
    # Iterate over each record in the dataset
    for record in output:
        # Check if the instrumentid matches any of the exchange instrument ids
        if (record['ExchangeNSEInstrumentId'] == instrumentid or
            record['ExchangeBSEInstrumentId'] == instrumentid or
            record['ExchangeMSEInstrumentId'] == instrumentid):
            
            # Return the holding quantity as an integer
            return record['HoldingQuantity']
    
    # Return None if no matching instrumentid is found
    return None

# Input from the user
instrument_id_input = int(input("Enter the instrumentid: "))

# Get the result
get_holding_quantity(instrument_id_input)
```

```python
def fetch_ltp(xtm,instruments):
    """
    Fetches LTP values from XTS API response with proper JSON parsing
    """
    try:
        # Unsubscribe first
        unsub_response = xtm.send_unsubscription(
            Instruments=instruments,
            xtsMessageCode=1502
        )
        
        if not unsub_response.get('result'):
            print(f"Unsubscription failed: {unsub_response}")
            return None

        # Then subscribe to get fresh data
        sub_response = xtm.send_subscription(
            Instruments=instruments,
            xtsMessageCode=1502
        )
        
        if not sub_response.get('result'):
            print(f"Subscription failed: {sub_response}")
            return None

        # Extract and parse JSON data from listQuotes
        
        for quote_json in sub_response['result']['listQuotes']:
            # Parse the JSON string
            quote_data = json.loads(quote_json)
            
            # Extract LTP from Touchline
            ltp = quote_data['Touchline']['LastTradedPrice']
            
        
        return ltp

    except KeyError as e:
        print(f"Missing expected key in response: {e}")
        return None
    except json.JSONDecodeError as e:
        print(f"Failed to parse JSON data: {e}")
        return None
    except Exception as e:
        print(f"Error fetching LTP: {str(e)}")
        return None
    
import json  # Make sure to import json module

# Sample usage
instruments = [{'exchangeSegment': 1, 'exchangeInstrumentID': 8080}]
ltp_values = fetch_ltp(xtm,instruments)

if ltp_values:
    print("Last Traded Prices:", ltp_values)
else:
    print("Failed to fetch LTP values")
```

```python
import pandas as pd
import time
import json
from threading import Lock

# Assuming xtm is a predefined API client
ltp_cache = {}
instrument_locks = {}  # Locks for each instrument to avoid race conditions

def get_instrument_details(symbol):
    """Get instrument details with TickSize default to 0.05 if missing"""
    try:
        response = xtm.get_equity_symbol(
            exchangeSegment=1,
            series='EQ',
            symbol=symbol
        )
        if response['result']:
            instrument_data = response['result'][0]
            price_band = instrument_data.get('PriceBand', {})
            return {
                'ExchangeInstrumentID': instrument_data['ExchangeInstrumentID'],
                'TickSize': instrument_data.get('TickSize', 0.05),
                'HighString': price_band.get('HighString', ''),
                'LowString': price_band.get('LowString', '')
            }
        return None
    except Exception as e:
        print(f"Error fetching details for {symbol}: {e}")
        return None

def fetch_current_ltp(exchange_instrument_id):
    """Fetch LTP with price adjustment logic based on high/low thresholds."""
    cache_expiry = 2  # seconds
    with instrument_locks.setdefault(exchange_instrument_id, Lock()):
        now = time.time()
        cached = ltp_cache.get(exchange_instrument_id)
        if cached and (now - cached['timestamp']) < cache_expiry:
            return cached['ltp']
        try:
            instruments = [{
                'exchangeSegment': 1, 
                'exchangeInstrumentID': int(exchange_instrument_id)
            }]
            xtm.send_unsubscription(Instruments=instruments, xtsMessageCode=1502)
            sub_response = xtm.send_subscription(Instruments=instruments, xtsMessageCode=1502)
            if not sub_response.get('result'):
                return None

            original_ltp = None
            for quote_json in sub_response['result']['listQuotes']:
                quote_data = json.loads(quote_json)
                original_ltp = quote_data['Touchline']['LastTradedPrice']
                break

            if original_ltp is None:
                return None

            ltp_cache[exchange_instrument_id] = {
                'ltp': original_ltp,
                'timestamp': time.time()
            }
            return original_ltp

        except Exception as e:
            print(f"LTP Error: {str(e)}")
            return None

def excel_to_trading_input(file_path):
    """Convert Excel input to trading system format."""
    master_instrument_df = pd.read_excel('master_instruments.xlsx')
    master_instrument_df['ISIN'] = master_instrument_df['ISIN'].str.strip()
    master_instrument_df = master_instrument_df.drop_duplicates(subset='ISIN', keep='first')
    master_instrument_df['instrumentID'] = master_instrument_df['instrumentID'].astype(str).str.strip()
    isin_to_details = master_instrument_df.set_index('ISIN').to_dict('index')
    
    df = pd.read_excel(file_path, sheet_name='Sheet1')
    df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
    
    stock_details = {}
    client_buy_details = {}
    client_sell_details = {}
    symbol_cache = {}

    def process_orders(order_df, is_buy=True):
        nonlocal stock_details
        details_dict = client_buy_details if is_buy else client_sell_details
        
        for _, row in order_df.iterrows():
            nse_symbol = row['NSE Symbol'].strip().upper()
            raw_isin = row['ISIN']
            isin = str(raw_isin).strip() if pd.notna(raw_isin) else ''
            
            if nse_symbol not in symbol_cache:
                api_details = get_instrument_details(nse_symbol)
                if api_details:
                    symbol_cache[nse_symbol] = api_details
                else:
                    isin_details = isin_to_details.get(isin, {})
                    symbol_cache[nse_symbol] = {
                        'ExchangeInstrumentID': str(isin_details.get('instrumentID', '')),
                        'TickSize': isin_details.get('TickSize', 0.05),
                        'HighString': isin_details.get('HighString', ''),
                        'LowString': isin_details.get('LowString', '')
                    }
                    if not symbol_cache[nse_symbol]['ExchangeInstrumentID']:
                        print(f"Warning: Couldn't resolve {nse_symbol}")
                        continue
            
            cached = symbol_cache.get(nse_symbol)
            if not cached or not cached['ExchangeInstrumentID']:
                continue
            
            exchange_id = int(cached['ExchangeInstrumentID'])
            client_id = row['USER ID']
            qty = int(row['QTY to Buy' if is_buy else 'QTY to Sell'])
            sov = int(row['SOV'])
            
            # Fetch and compare with LTP
            current_ltp = fetch_current_ltp(exchange_id)
            if current_ltp is None or sov < current_ltp:
                print(f"Skipping {nse_symbol} (ID: {exchange_id}) for client {client_id}: SOV {sov} < LTP {current_ltp}")
                continue
            
            if exchange_id not in stock_details:
                stock_details[exchange_id] = {
                    'name': nse_symbol,
                    'exchange_instrument_id': exchange_id,
                    'tick_size': cached.get('TickSize', 0.05),
                    'high_string': cached['HighString'],
                    'low_string': cached['LowString']
                }
            
            if client_id not in details_dict:
                details_dict[client_id] = {
                    'single_order_value': sov,
                    f'total_quantity_to_{"buy" if is_buy else "sell"}': {}
                }
            
            qty_key = f'total_quantity_to_{"buy" if is_buy else "sell"}'
            current = details_dict[client_id][qty_key].get(exchange_id, 0)
            details_dict[client_id][qty_key][exchange_id] = current + qty

    process_orders(df[df['QTY to Buy'].notna()], is_buy=True)
    process_orders(df[df['QTY to Sell'].notna()], is_buy=False)
    
    all_exchange_ids = set()
    for client_data in client_buy_details.values():
        all_exchange_ids.update(client_data['total_quantity_to_buy'].keys())
    for client_data in client_sell_details.values():
        all_exchange_ids.update(client_data['total_quantity_to_sell'].keys())
    
    stock_details = {exch_id: details for exch_id, details in stock_details.items() if exch_id in all_exchange_ids}
    return stock_details, client_buy_details, client_sell_details

if __name__ == "__main__":
    file_path = "Sample Excel Inputalg.xlsx"
    stock_details, client_buy_details, client_sell_details = excel_to_trading_input(file_path)
    print("Instrument Details:", stock_details)
    print("\nBuy Orders:", client_buy_details)
    print("\nSell Orders:", client_sell_details)
```

```python
import time
import json
import os
import threading
from threading import Event, Thread
from datetime import datetime
from collections import defaultdict

# -------------------------
# Global Variables & Setup
# -------------------------

# Lock and cache for instrument LTP calls
instrument_locks = defaultdict(threading.Lock)
ltp_cache = {}

# -------------------------
# API Initialization
# -------------------------
# Replace these with your actual API instances.
# (Initialize your API instances as needed)
xtm = xtm  # Market Data API instance
client_apis = {
    "ALG347": xti,      # Replace with actual API instance for client ALG347
    #"Z46066": xti_zeb,  # Replace with actual API instance for client Z46066
    #"ADV21":  xti2,     # Replace with actual API instance for client ADV21
    #"ADV26":  xti3,     # Replace with actual API instance for client ADV26
    #"ALG330": xti4,
    #"MRG1":xti5,
    ###"SBJ01":xti6,
    #"SBJ20":xti7,
    #"SBJ22":xti8,
    #'ALG303':xti9,
    #'ALG337':xti10,
    #'ALG345':xti11,
    #'ALG328':xti12,
    #'SBJ15':xti13,
    #'ADV03':xti14,
    ###'ADV06':xti15,
    #'ADV27':xti16,
    #'ALG301':xti17,
    #'SBJ04':xti18,
    #'ALG300':xti19,

    
    # Replace with actual API instance for client ALG330
    # ... add all 25 clients as needed
}

# -------------------------
# Configuration
# -------------------------
# Replace these with your actual details.
stock_details = stock_details
client_details = client_buy_details

# -------------------------
# Function Definitions
# -------------------------

def fetch_current_ltp(exchange_instrument_id):
    """Fetch current LTP with caching and price adjustment based on tick size and thresholds."""
    cache_expiry = 2  # seconds
    with instrument_locks[exchange_instrument_id]:
        now = time.time()
        cached = ltp_cache.get(exchange_instrument_id)
        if cached and (now - cached['timestamp']) < cache_expiry:
            return cached['ltp']
        try:
            instruments = [{
                'exchangeSegment': 1,
                'exchangeInstrumentID': int(exchange_instrument_id)
            }]
            # Unsubscribe and subscribe to refresh market data
            xtm.send_unsubscription(Instruments=instruments, xtsMessageCode=1502)
            sub_response = xtm.send_subscription(Instruments=instruments, xtsMessageCode=1502)
            if not sub_response.get('result'):
                return None

            original_ltp = None
            for quote_json in sub_response['result']['listQuotes']:
                quote_data = json.loads(quote_json)
                original_ltp = quote_data['Touchline']['LastTradedPrice']
                break
            if original_ltp is None:
                return None

            ex_id_int = int(exchange_instrument_id)
            stock_info = stock_details.get(ex_id_int, {})
            tick_size = stock_info.get('tick_size', 0.05)
            high = float(stock_info.get('high_string', 0))
            low = float(stock_info.get('low_string', 0))

            # Adjust LTP to 99.5% and round to the tick size
            adjusted_ltp = original_ltp
            adjusted_ltp = round(adjusted_ltp / tick_size) * tick_size

            if high > 0 and low > 0:
                if adjusted_ltp >= high or adjusted_ltp <= low:
                    final_ltp = round(original_ltp / tick_size) * tick_size
                else:
                    final_ltp = adjusted_ltp
            else:
                final_ltp = adjusted_ltp

            ltp_cache[exchange_instrument_id] = {'ltp': final_ltp, 'timestamp': time.time()}
            return final_ltp

        except Exception as e:
            print(f"LTP Error: {str(e)}")
            return None

def calculate_order_parameters(single_order_value, total_quantity_to_buy, ltp):
    """Calculate order quantity for a single order."""
    if ltp == 0 or single_order_value == 0 or total_quantity_to_buy == 0:
        raise ValueError("Invalid parameters: Values cannot be zero.")
    single_order_quantity = int(single_order_value / ltp)
    return max(single_order_quantity, 1)

def get_positions_by_day(api, client_id):
    """Fetch day-wise positions for a client."""
    try:
        response = api.get_position_daywise(clientID=client_id)
        if response.get('type') == 'success':
            positions = response['result'].get('positionList', [])
            return [{'TradingSymbol': pos['TradingSymbol'], 'Quantity': pos['Quantity']} for pos in positions]
        else:
            print(f"Error fetching positions for {client_id}: {response.get('message')}")
            return []
    except Exception as e:
        print(f"Position check error for {client_id}: {e}")
        return []

def place_order(client_id, exchange_instrument_id, ltp, quantity, order_responses, lock):
    """Place a buy order using the client-specific API."""
    try:
        api = client_apis.get(client_id)
        if api is None:
            print(f"No API found for {client_id}")
            return 0

        response = api.place_order(
            exchangeSegment=api.EXCHANGE_NSECM,
            exchangeInstrumentID=exchange_instrument_id,
            productType=api.PRODUCT_CNC,
            orderType=api.ORDER_TYPE_LIMIT,
            orderSide=api.TRANSACTION_TYPE_BUY,
            timeInForce=api.VALIDITY_DAY,
            disclosedQuantity=0,
            orderQuantity=quantity,
            limitPrice=ltp,
            stopPrice=0,
            orderUniqueIdentifier=str(datetime.now().timestamp()),
            clientID=client_id
        )

        if response['type'] != 'error':
            order_id = response['result']['AppOrderID']
            with lock:
                order_responses.setdefault((exchange_instrument_id, client_id), []).append({
                    'order_id': order_id,
                    'quantity': quantity,
                    'client_id': client_id
                })
            print(f"[{datetime.now()}] {client_id} Order {order_id}: {quantity} @ {ltp}")
            return quantity
        print(f"Order failed {client_id}: {response.get('message')}")
        return 0
    except Exception as e:
        print(f"Order Error {client_id}: {str(e)}")
        return 0

def order_scheduler(client_id, exchange_instrument_id, details, client_params, total_time, order_responses, lock):
    """Schedules and places orders for a given client and instrument."""
    total_quantity = client_params['total_quantity_to_buy'].get(exchange_instrument_id, 0)
    if total_quantity <= 0:
        return

    ltp = fetch_current_ltp(exchange_instrument_id)
    if not ltp:
        return

    try:
        single_order_qty = max(1, int(client_params['single_order_value'] / ltp))
        n_orders = max(1, (total_quantity + single_order_qty - 1) // single_order_qty)
        interval = total_time / n_orders
    except Exception as e:
        print(f"Calculation Error {client_id}: {str(e)}")
        return

    print(f"\n{client_id} Starting {details.get('name', exchange_instrument_id)}:")
    print(f"Total: {total_quantity}, Orders: {n_orders}, Interval: {interval:.2f}s")

    remaining = total_quantity
    start_time = time.time()
    for _ in range(n_orders):
        if remaining <= 0:
            break
        current_ltp = fetch_current_ltp(exchange_instrument_id)
        executed = place_order(
            client_id, exchange_instrument_id, current_ltp,
            min(single_order_qty, remaining), order_responses, lock
        )
        remaining -= executed
        time.sleep(max(0, interval - (time.time() - start_time) % interval))

def place_orders_with_timing(stock_details, client_details, total_time):
    """Starts order scheduling threads for all client/instrument pairs."""
    order_responses = {}
    lock = threading.Lock()
    threads = []

    for client_id, client_params in client_details.items():
        for ex_id, details in stock_details.items():
            total_quantity = client_params['total_quantity_to_buy'].get(ex_id, 0)
            if total_quantity <= 0:
                continue
            thread = threading.Thread(
                target=order_scheduler,
                args=(client_id, ex_id, details, client_params, total_time, order_responses, lock)
            )
            threads.append(thread)
            thread.start()

    for thread in threads:
        thread.join()

    return order_responses

def check_order_status(api, client_id, order_id):
    """Check order status via client-specific API."""
    try:
        order_book = api.get_order_book(clientID=client_id)
        return any(str(o['AppOrderID']) == str(order_id) and 
                   o['OrderStatus'].lower() not in ['filled', 'rejected']
                   for o in order_book.get('result', []))
    except Exception as e:
        print(f"Status Error {client_id}: {str(e)}")
        return False

def modify_orders(order_responses):
    """Modify orders if they remain open."""
    for (ex_id, client_id), orders in order_responses.items():
        api = client_apis.get(client_id)
        if not api:
            continue
        for order_info in orders:
            if check_order_status(api, client_id, order_info['order_id']):
                current_ltp = fetch_current_ltp(ex_id)
                if current_ltp:
                    response = api.modify_order(
                        appOrderID=order_info['order_id'],
                        modifiedProductType=api.PRODUCT_CNC,
                        modifiedOrderType=api.ORDER_TYPE_LIMIT,
                        modifiedOrderQuantity=order_info['quantity'],
                        modifiedDisclosedQuantity=0,
                        modifiedLimitPrice=current_ltp,
                        modifiedStopPrice=0,
                        modifiedTimeInForce=api.VALIDITY_DAY,
                        orderUniqueIdentifier=str(datetime.now().timestamp()),
                        clientID=client_id
                    )
                    print(f"{client_id} Modified {order_info['order_id']} @ {current_ltp}")

def get_client_orders(api, client_id):
    """Fetch client's order book and separate open and rejected orders."""
    try:
        order_response = api.get_order_book(clientID=client_id)
        if order_response.get('type') != 'success':
            return [], []
        
        orders = order_response.get('result', [])
        open_orders = []
        rejected_orders = []
        
        for order in orders:
            if order['OrderStatus'] in ['New', 'Replaced']:
                open_orders.append(order)
            elif order['OrderStatus'] == 'Rejected':
                rejected_orders.append({
                    'symbol': order['TradingSymbol'],
                    'order_id': order.get('AppOrderID', 'N/A'),
                    'timestamp': order.get('OrderGeneratedDateTime', 'N/A'),
                    'reason': order.get('CancelRejectReason', 'No reason provided')
                })
        
        return open_orders, rejected_orders
    except Exception as e:
        print(f"Order fetch error for {client_id}: {str(e)}")
        return [], []

def position_monitor(client_details, stop_event):
    """
    Monitors and logs live positions, open orders, and rejected orders
    for each client into a log file (with date & time details).
    """
    while not stop_event.is_set():
        for client_id, params in client_details.items():
            try:
                api = client_apis.get(client_id)
                if not api:
                    continue
                
                log_entry = f"\n[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] Client: {client_id}\n"
                
                # Fetch positions
                positions = api.get_position_daywise(clientID=client_id)
                if positions.get('type') == 'success':
                    log_entry += "Positions:\n"
                    for p in positions['result'].get('positionList', []):
                        log_entry += f"  {p['TradingSymbol']}: {p['Quantity']}\n"
                else:
                    log_entry += "Position fetch error\n"
                
                # Fetch orders
                open_orders, rejected_orders = get_client_orders(api, client_id)
                log_entry += "\nOpen Orders:\n"
                if open_orders:
                    for o in open_orders:
                        log_entry += f"  {o['TradingSymbol']} - Qty: {o['OrderQuantity']} (Status: {o['OrderStatus']})\n"
                else:
                    log_entry += "  No open orders\n"
                
                log_entry += "\nRejected Orders:\n"
                if rejected_orders:
                    for r in rejected_orders:
                        log_entry += f"  {r['symbol']} - Reason: {r['reason']} (Time: {r['timestamp']})\n"
                else:
                    log_entry += "  No rejected orders\n"
                
                with open(f"positions_{client_id}.log", "a") as f:
                    f.write(log_entry + "\n")
            
            except Exception as e:
                error_msg = f"Monitoring error {client_id}: {str(e)}"
                with open(f"positions_{client_id}.log", "a") as f:
                    f.write(error_msg + "\n")
        
        time.sleep(10)

def generate_text_summary(client_details):
    """Generate a comprehensive text summary report (positions, orders, rejections)."""
    if not os.path.exists('reports'):
        os.makedirs('reports')
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"reports/trading_summary_{timestamp}.txt"
    
    report_content = []
    report_content.append("=" * 60)
    report_content.append(f"TRADING SUMMARY REPORT - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    report_content.append("=" * 60 + "\n")
    
    total_clients = 0
    total_positions = 0
    total_open_orders = 0
    total_rejections = 0
    
    for client_id in client_details:
        try:
            api = client_apis.get(client_id)
            if not api:
                continue
            
            client_section = []
            client_section.append(f"\nCLIENT: {client_id}")
            client_section.append("-" * 60)
            
            positions = []
            pos_response = api.get_position_daywise(clientID=client_id)
            if pos_response.get('type') == 'success':
                positions = pos_response['result'].get('positionList', [])
            
            open_orders, rejected_orders = get_client_orders(api, client_id)
            
            client_section.append("\nCURRENT POSITIONS:")
            if positions:
                for pos in positions:
                    client_section.append(
                        f"  {pos['TradingSymbol']:<15} | Quantity: {pos['Quantity']:<6} | Avg Price: {pos.get('AveragePrice', 'N/A')}"
                    )
                total_positions += len(positions)
            else:
                client_section.append("  No positions found")
            
            client_section.append("\nOPEN ORDERS:")
            if open_orders:
                for order in open_orders:
                    client_section.append(
                        f"  {order['TradingSymbol']:<15} | ID: {order['AppOrderID']:<12} | Qty: {order['OrderQuantity']:<5} | Price: {order['OrderPrice']:<8} | Status: {order['OrderStatus']}"
                    )
                total_open_orders += len(open_orders)
            else:
                client_section.append("  No open orders")
            
            client_section.append("\nREJECTED ORDERS:")
            if rejected_orders:
                for reject in rejected_orders:
                    try:
                        ts = datetime.strptime(reject['timestamp'], '%d-%m-%Y %H:%M:%S')
                        formatted_ts = ts.strftime('%Y-%m-%d %H:%M:%S')
                    except:
                        formatted_ts = reject['timestamp']
                    
                    client_section.append(
                        f"  {formatted_ts} | ID: {reject['order_id']:<12} | {reject['symbol']:<15} | Reason: {reject['reason'][:55]}..."
                    )
                total_rejections += len(rejected_orders)
            else:
                client_section.append("  No rejected orders")
            
            report_content.append('\n'.join(client_section))
            total_clients += 1
            
        except Exception as e:
            report_content.append(f"\nError processing {client_id}: {str(e)}")
    
    report_content.append("\n" + "=" * 60)
    report_content.append("SUMMARY STATISTICS")
    report_content.append("-" * 60)
    report_content.append(f"Total Clients Processed: {total_clients}")
    report_content.append(f"Total Positions Found:   {total_positions}")
    report_content.append(f"Total Open Orders:       {total_open_orders}")
    report_content.append(f"Total Rejected Orders:   {total_rejections}")
    report_content.append("=" * 60)
    
    with open(filename, 'w') as f:
        f.write('\n'.join(report_content))
    
    print(f"Report generated: {filename}")
    return filename

def parse_order_time(ts_str):
    """Parse order timestamp using multiple potential formats."""
    for fmt in (
        '%Y-%m-%d %H:%M:%S.%f',  # With microseconds
        '%Y-%m-%d %H:%M:%S',     # Without microseconds
        '%d-%m-%Y %H:%M:%S',     # Day-month-year format
        '%m/%d/%Y %H:%M:%S',     # Month/day/year
        '%d-%b-%Y %H:%M:%S',     # With abbreviated month
    ):
        try:
            return datetime.strptime(ts_str, fmt)
        except ValueError:
            continue
    print(f"Warning: Failed to parse timestamp '{ts_str}'")
    return None

def display_final_summary_buy(client_details, order_responses, order_start_time):
    """
    Display a final summary report including orders placed after the strategy start time
    and live positions. Only orders with status exactly 'FILLED' are counted as executed.
    """
    print("\n==================== Final Summary Report ====================")
    for client_id, params in client_details.items():
        print(f"\nClient: {client_id}")
        print("-" * 70)
        header = f"{'Instrument':<15} | {'Target':>6} | {'Executed':>8} | {'Open':>6} | {'Rejected':>8} | {'Live Pos':>8}"
        print(header)
        print("-" * len(header))
        api = client_apis.get(client_id)
        if not api:
            continue

        try:
            order_book_response = api.get_order_book(clientID=client_id)
            pos_response = api.get_position_daywise(clientID=client_id)
        except Exception as e:
            print(f"Error fetching data for {client_id}: {str(e)}")
            continue

        live_positions = {}
        if pos_response.get('type') == 'success':
            for pos in pos_response['result'].get('positionList', []):
                trading_symbol = pos.get('TradingSymbol', 'N/A')
                quantity = pos.get('Quantity', 0)
                inst_id = pos.get('ExchangeInstrumentID') or pos.get('ExchangeNSEInstrumentID') or pos.get('ExchangeNSEInstrumentId')
                if inst_id:
                    try:
                        inst_id = int(inst_id)
                        live_positions[inst_id] = {'trading_symbol': trading_symbol, 'quantity': quantity}
                    except Exception as e:
                        print(f"Error processing live position for {client_id}: {e}")
                        continue

        instrument_orders = defaultdict(lambda: {'executed': 0, 'open': 0, 'rejected': 0})
        orders = order_book_response.get('result', []) if order_book_response.get('type') == 'success' else []
        for order in orders:
            try:
                order_inst_id = order.get('ExchangeInstrumentID') or order.get('ExchangeNSEInstrumentID') or order.get('ExchangeNSEInstrumentId')
                if not order_inst_id:
                    continue
                order_inst_id = int(order_inst_id)
                status = order.get('OrderStatus', '').upper()
                order_time = parse_order_time(order.get('OrderGeneratedDateTime', ''))
                if not order_time or order_time < order_start_time:
                    continue
                order_qty = int(order.get('OrderQuantity', 0))
                if status == "FILLED":
                    instrument_orders[order_inst_id]['executed'] += order_qty
                elif status in ["NEW", "REPLACED", "OPEN", "PARTIALLY_FILLED"]:
                    instrument_orders[order_inst_id]['open'] += order_qty
                elif status in ["REJECTED","CANCELLED"]:
                    instrument_orders[order_inst_id]['rejected'] += order_qty
            except Exception as e:
                print(f"Error processing order for {client_id}: {str(e)}")

        # For buy orders, we use 'total_quantity_to_buy'
        for ex_id, target_qty in params.get('total_quantity_to_buy', {}).items():
            try:
                ex_id_int = int(ex_id)
            except:
                ex_id_int = ex_id
            stock_name = stock_details.get(ex_id_int, {}).get('name', f"ID:{ex_id}")
            orders_data = instrument_orders.get(ex_id_int, {'executed': 0, 'open': 0, 'rejected': 0})
            live_qty = live_positions.get(ex_id_int, {}).get('quantity', 0)
            print(f"{stock_name:<15} | {target_qty:6} | {orders_data['executed']:8} | {orders_data['open']:6} | {orders_data['rejected']:8} | {live_qty:8}")

        print("\nDetailed Live Positions:")
        if live_positions:
            for inst_id, info in live_positions.items():
                print(f"  {info['trading_symbol']}: {info['quantity']} shares")
        else:
            print("  No live positions available.")

# -------------------------
# Main Execution
# -------------------------

if __name__ == "__main__":
    # (Re)initialize APIs if needed here.
    xtm = xtm  # Your market data API initialization
    client_apis = client_apis
    
    order_responses = {}
    lock = threading.Lock()
    total_time = 5 * 60  # For example, 5 minutes for order placement
    
    # Start the monitoring thread (logs live positions/orders to file)
    stop_event = Event()
    monitor_thread = Thread(target=position_monitor, args=(client_details, stop_event))
    monitor_thread.daemon = True
    monitor_thread.start()
    
    order_start_time = datetime.now()
    threads = []
    for client_id, params in client_details.items():
        for ex_id, total_qty in params.get('total_quantity_to_buy', {}).items():
            if ex_id not in stock_details:
                continue
            thread = threading.Thread(
                target=order_scheduler,
                args=(client_id, ex_id, stock_details[ex_id], params, total_time, order_responses, lock)
            )
            threads.append(thread)
            thread.start()
    
    for t in threads:
        t.join()
    
    # Optionally modify orders a few times at intervals
    for _ in range(20):
        modify_orders(order_responses)
        time.sleep(30)
    
    # Stop the monitoring thread
    stop_event.set()
    monitor_thread.join()
    
    # Generate the final text report
    #report_path = generate_text_summary(client_details)
    #print(f"\nFinal report saved to: {report_path}")
    
    # Display the final summary report on the console
    #display_final_summary_buy(client_details, order_responses, order_start_time)
```

```python
def cancel_all_open_orders(stock_details, client_details):
    """
    Cancel all open orders for all stocks across all clients.
    
    Args:
        stock_details (dict): Dictionary containing stock instrument details.
        client_details (dict): Dictionary containing client configurations.
        
    Returns:
        dict: Response details for each cancellation request.
    """
    cancellation_responses = {}
    
    for client_id, client_info in client_details.items():
        # Get the appropriate API instance for the client from client_apis dictionary.
        client_api = client_apis.get(client_id)
        if client_api is None:
            print(f"No API instance found for client {client_id}")
            continue
        
        cancellation_responses[client_id] = {}
        
        for ticker, stock_info in stock_details.items():
            try:
                # Get exchange instrument ID from stock details.
                exchange_instrument_id = stock_info['exchange_instrument_id']
                
                # Send cancellation request.
                response = client_api.cancelall_order(
                    exchangeInstrumentID=exchange_instrument_id,
                    exchangeSegment=client_api.EXCHANGE_NSECM
                )
                
                cancellation_responses[client_id][ticker] = {
                    'success': response['type'] != 'error',
                    'message': response['message'] if response['type'] == 'error' else 'Cancellation successful',
                    'response': response
                }
                
            except KeyError as e:
                cancellation_responses[client_id][ticker] = {
                    'success': False,
                    'message': f"Missing key in stock details: {str(e)}"
                }
            except Exception as e:
                cancellation_responses[client_id][ticker] = {
                    'success': False,
                    'message': f"Error cancelling orders: {str(e)}"
                }
    
    return cancellation_responses

# Cancel all orders and print results
cancellation_results = cancel_all_open_orders(stock_details, client_details)

# Print cancellation results for all clients
for client_id, stock_responses in cancellation_results.items():
    print(f"\nCancellation results for {client_id}:")
    for ticker, response in stock_responses.items():
        status = "Success" if response['success'] else "Failed"
        print(f"{ticker}: {status} - {response['message']}")
```

```python
### import time
import json
import os
import threading
from threading import Event, Thread
from datetime import datetime
from collections import defaultdict

# ------------------------------
# Market Data API & Client APIs
# ------------------------------
# (Initialize your API instances as needed)
xtm = xtm  # Market Data API instance
client_apis = {
    "ALG347": xti,      # Replace with actual API instance for client ALG347
    #"Z46066": xti_zeb,  # Replace with actual API instance for client Z46066
    #"ADV21":  xti2,     # Replace with actual API instance for client ADV21
    #"ADV26":  xti3,     # Replace with actual API instance for client ADV26
    #"ALG330": xti4,
    #"MRG1":xti5,
    #"SBJ01":xti6,
    #"SBJ20":xti7,
    #"SBJ22":xti8,
    #'ALG303':xti9,
    #'ALG337':xti10,
    #'ALG345':xti11,
    #'ALG328':xti12,
    #'SBJ15':xti13,
    #'ADV03':xti14,
    ###'ADV06':xti15,
    #'ADV27':xti16,
    #'ALG301':xti17,
    #'SBJ04':xti18,
    #'ALG300':xti19,

    
    # Replace with actual API instance for client ALG330
    # ... add all 25 clients as needed
}

# ------------------------------
# Global variables for instrument locks and LTP cache
# ------------------------------
instrument_locks = defaultdict(threading.Lock)
ltp_cache = {}

def fetch_current_ltp(exchange_instrument_id):
    """Fetch LTP with price adjustment logic based on high/low thresholds."""
    cache_expiry = 2  # seconds
    with instrument_locks[exchange_instrument_id]:
        now = time.time()
        cached = ltp_cache.get(exchange_instrument_id)
        if cached and (now - cached['timestamp']) < cache_expiry:
            return cached['ltp']
        try:
            instruments = [{
                'exchangeSegment': 1, 
                'exchangeInstrumentID': int(exchange_instrument_id)
            }]
            # Market data subscription logic
            unsub_response = xtm.send_unsubscription(Instruments=instruments, xtsMessageCode=1502)
            sub_response = xtm.send_subscription(Instruments=instruments, xtsMessageCode=1502)
            if not sub_response.get('result'):
                return None

            original_ltp = None
            for quote_json in sub_response['result']['listQuotes']:
                quote_data = json.loads(quote_json)
                original_ltp = quote_data['Touchline']['LastTradedPrice']
                break

            if original_ltp is None:
                return None

            # Get stock parameters
            ex_id_int = int(exchange_instrument_id)
            stock_info = stock_details.get(ex_id_int, {})
            tick_size = stock_info.get('tick_size', 0.05)
            high = float(stock_info.get('high_string', 0))
            low = float(stock_info.get('low_string', 0))

            # Calculate adjusted LTP(LTP +-%) 
            adjusted_ltp = original_ltp
            adjusted_ltp = round(adjusted_ltp / tick_size) * tick_size

            if high > 0 and low > 0:
                if adjusted_ltp >= high or adjusted_ltp <= low:
                    final_ltp = round(original_ltp / tick_size) * tick_size
                else:
                    final_ltp = adjusted_ltp
            else:
                final_ltp = adjusted_ltp

            ltp_cache[exchange_instrument_id] = {
                'ltp': final_ltp,
                'timestamp': time.time()
            }
            return final_ltp

        except Exception as e:
            print(f"LTP Error: {str(e)}")
            return None

def fetch_sell_ltp(exchange_instrument_id):
    """Fetch LTP for sell orders with tick rounding."""
    base_ltp = fetch_current_ltp(exchange_instrument_id)
    if base_ltp is None:
        return None
    ex_id_int = int(exchange_instrument_id)
    tick_size = stock_details.get(ex_id_int, {}).get('tick_size', 0.05)
    adjusted_ltp = round(base_ltp / tick_size) * tick_size
    return adjusted_ltp

# ------------------------------
# Holding and Order Functions
# ------------------------------
def get_holding_quantity(client_id, exchange_instrument_id, api):
    """Fetch holding quantity for a specific instrument and client."""
    try:
        response = api.get_holding(clientID=client_id)
        if response['type'] != 'success':
            print(f"Error fetching holdings for {client_id}: {response.get('message')}")
            return None
        holdings = response['result'].get('RMSHoldings', {}).get('Holdings', {})
        for isin, details in holdings.items():
            if details.get('ExchangeNSEInstrumentId') == exchange_instrument_id:
                return int(details.get('HoldingQuantity', 0))
        return 0
    except Exception as e:
        print(f"Exception in get_holding_quantity for {client_id}: {e}")
        return None

def validate_holding_quantity(client_id, exchange_instrument_id, total_quantity_to_sell, api):
    """Validate if client has enough holdings to sell considering open orders."""
    holding_qty = get_holding_quantity(client_id, exchange_instrument_id, api)
    if holding_qty is None:
        print(f"Validation failed: Could not fetch holdings for {client_id}")
        return False

    open_sell_qty = 0
    try:
        order_response = api.get_order_book(clientID=client_id)
        if order_response.get('type') == 'success':
            for order in order_response['result']:
                order_instrument_id = str(order.get('ExchangeInstrumentID', ''))
                target_instrument_id = str(exchange_instrument_id)
                if order_instrument_id == target_instrument_id:
                    if order.get('OrderSide', '').upper() == 'SELL':
                        status = order.get('OrderStatus', '').upper()
                        if status in ['NEW', 'REPLACED', 'PARTIALLY_FILLED', 'OPEN']:
                            open_sell_qty += order.get('OrderQuantity', 0)
    except Exception as e:
        print(f"Error fetching open orders for {client_id}: {str(e)}")
        return False

    available_qty = holding_qty - open_sell_qty
    if available_qty < total_quantity_to_sell:
        print(f"Validation failed: {client_id} has {holding_qty} units, {open_sell_qty} in open orders. Available: {available_qty}, needed: {total_quantity_to_sell}")
        return False
    return True

def place_sell_order(client_id, exchange_instrument_id, ltp, quantity, order_responses, lock):
    """Place sell order using client-specific API."""
    try:
        api = client_apis.get(client_id)
        if api is None:
            print(f"No API found for {client_id}")
            return 0
        response = api.place_order(
            exchangeSegment=api.EXCHANGE_NSECM,
            exchangeInstrumentID=exchange_instrument_id,
            productType=api.PRODUCT_CNC,
            orderType=api.ORDER_TYPE_LIMIT,
            orderSide=api.TRANSACTION_TYPE_SELL,
            timeInForce=api.VALIDITY_DAY,
            disclosedQuantity=0,
            orderQuantity=quantity,
            limitPrice=ltp,
            stopPrice=0,
            orderUniqueIdentifier=str(datetime.now().timestamp()),
            clientID=client_id
        )
        if response['type'] != 'error':
            order_id = response['result']['AppOrderID']
            with lock:
                order_responses.setdefault((exchange_instrument_id, client_id), []).append({
                    'order_id': order_id,
                    'quantity': quantity,
                    'client_id': client_id
                })
            print(f"[{datetime.now()}] {client_id} Sell Order {order_id}: {quantity} @ {ltp}")
            return quantity
        print(f"Sell order failed {client_id}: {response.get('message')}")
        return 0
    except Exception as e:
        print(f"Sell Order Error {client_id}: {str(e)}")
        return 0

def schedule_sell_orders(client_id, exchange_instrument_id, details, params, total_time, order_responses, lock):
    """Schedule sell orders for a client and instrument."""
    api = client_apis.get(client_id)
    if not api:
        return

    total_quantity = params['total_quantity_to_sell'].get(exchange_instrument_id, 0)
    if total_quantity <= 0:
        return

    if not validate_holding_quantity(client_id, int(exchange_instrument_id), total_quantity, api):
        return

    ltp = fetch_sell_ltp(exchange_instrument_id)
    if not ltp:
        return

    try:
        single_order_qty = max(1, int(params['single_order_value'] / ltp))
        n_orders = max(1, (total_quantity + single_order_qty - 1) // single_order_qty)
        interval = total_time / n_orders
    except Exception as e:
        print(f"Calculation Error {client_id}: {str(e)}")
        return

    print(f"\n{client_id} Selling {details['name']}:")
    print(f"Total: {total_quantity}, Orders: {n_orders}, Interval: {interval:.2f}s")

    remaining = total_quantity
    start_time = time.time()
    for _ in range(n_orders):
        if remaining <= 0:
            break
        current_ltp = fetch_sell_ltp(exchange_instrument_id)
        if not current_ltp:
            continue
        qty = min(single_order_qty, remaining)
        executed = place_sell_order(client_id, exchange_instrument_id, current_ltp, qty, order_responses, lock)
        remaining -= executed
        elapsed = time.time() - start_time
        sleep_time = max(0, interval - (elapsed % interval))
        time.sleep(sleep_time)

def check_order_status(api, client_id, order_id):
    """Check order status from the order book using the order ID."""
    try:
        response = api.get_order_book(clientID=client_id)
        if response.get('type') == 'success':
            for order in response['result']:
                if order.get('AppOrderID') == order_id:
                    status = order.get('OrderStatus', '').upper()
                    return status in ['NEW', 'REPLACED', 'PARTIALLY_FILLED', 'OPEN']
        return False
    except Exception as e:
        print(f"Status check failed for {order_id}: {str(e)}")
        return False

def modify_sell_orders(order_responses):
    """Modify live orders by checking their status and updating the limit price if needed.
       If no order qualifies for modification, immediately prints a message indicating
       that there are no orders left to modify.
    """
    overall_modified = False  # Flag to track any modification overall
    # Iterate through each (exchange instrument, client) pair in order_responses.
    for (ex_id, client_id), orders in order_responses.items():
        api = client_apis.get(client_id)
        if not api:
            continue

        modified_for_pair = False  # Flag for current (ex_id, client_id) pair
        for order_info in orders:
            if check_order_status(api, client_id, order_info['order_id']):
                current_ltp = fetch_sell_ltp(ex_id)
                if current_ltp:
                    response = api.modify_order(
                        appOrderID=order_info['order_id'],
                        modifiedProductType=api.PRODUCT_CNC,
                        modifiedOrderType=api.ORDER_TYPE_LIMIT,
                        modifiedOrderQuantity=order_info['quantity'],
                        modifiedDisclosedQuantity=0,
                        modifiedLimitPrice=current_ltp,
                        modifiedStopPrice=0,
                        modifiedTimeInForce=api.VALIDITY_DAY,
                        orderUniqueIdentifier=str(datetime.now().timestamp()),
                        clientID=client_id
                    )
                    print(f"{client_id} Modified sell order {order_info['order_id']} @ {current_ltp}")
                    overall_modified = True
                    modified_for_pair = True
        if not modified_for_pair:
            print(f"{client_id} - For instrument {ex_id}: No order Left to modify based on OrderID.")
    if not overall_modified:
        print("No order Left to modify based on OrderID.")

# ------------------------------
# Strategy Execution
# ------------------------------
def execute_sell_strategy(stock_details, client_details, total_time):
    """Orchestrate sell order placement across clients and stocks."""
    order_responses = {}
    lock = threading.Lock()
    threads = []
    for client_id, params in client_details.items():
        for ex_id, details in stock_details.items():
            if ex_id not in params.get('total_quantity_to_sell', {}):
                continue
            thread = threading.Thread(
                target=schedule_sell_orders,
                args=(client_id, ex_id, details, params, total_time, order_responses, lock)
            )
            threads.append(thread)
            thread.start()
    for thread in threads:
        thread.join()
    return order_responses

# ------------------------------
# Order Fetching and Reporting Helpers
# ------------------------------
def fetch_client_orders(api, client_id):
    """
    Fetch all orders for a given client via the API.
    Returns open orders and rejected orders.
    """
    try:
        order_response = api.get_order_book(clientID=client_id)
        if order_response.get('type') != 'success':
            return [], []
        orders = order_response.get('result', [])
        open_orders = []
        rejected_orders = []
        for order in orders:
            if order['OrderStatus'] in ['New', 'Replaced']:
                open_orders.append(order)
            elif order['OrderStatus'] == 'Rejected':
                rejected_orders.append({
                    'symbol': order['TradingSymbol'],
                    'order_id': order.get('AppOrderID', 'N/A'),
                    'timestamp': order.get('OrderGeneratedDateTime', 'N/A'),
                    'reason': order.get('CancelRejectReason', 'No reason provided')
                })
        return open_orders, rejected_orders
    except Exception as e:
        print(f"Order fetch error for {client_id}: {str(e)}")
        return [], []

def generate_text_summary(client_details):
    """Generate a comprehensive text report with positions, orders, and rejections."""
    if not os.path.exists('reports'):
        os.makedirs('reports')
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"reports/trading_summary_{timestamp}.txt"
    report_content = []
    report_content.append("=" * 60)
    report_content.append(f"TRADING SUMMARY REPORT - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    report_content.append("=" * 60 + "\n")
    total_clients = 0
    total_positions = 0
    total_open_orders = 0
    total_rejections = 0
    for client_id in client_details:
        try:
            api = client_apis.get(client_id)
            if not api:
                continue
            client_section = []
            client_section.append(f"\nCLIENT: {client_id}")
            client_section.append("-" * 60)
            # Positions
            positions = []
            pos_response = api.get_position_daywise(clientID=client_id)
            if pos_response.get('type') == 'success':
                positions = pos_response['result'].get('positionList', [])
            client_section.append("\nCURRENT POSITIONS:")
            if positions:
                for pos in positions:
                    client_section.append(
                        f"  {pos['TradingSymbol']:<15} | Quantity: {pos['Quantity']:<6} | Avg Price: {pos.get('AveragePrice', 'N/A')}"
                    )
                total_positions += len(positions)
            else:
                client_section.append("  No positions found")
            # Open and rejected orders
            open_orders, rejected_orders = fetch_client_orders(api, client_id)
            client_section.append("\nOPEN ORDERS:")
            if open_orders:
                for order in open_orders:
                    client_section.append(
                        f"  {order['TradingSymbol']:<15} | ID: {order['AppOrderID']:<12} | Qty: {order['OrderQuantity']:<5} | Price: {order['OrderPrice']:<8} | Status: {order['OrderStatus']}"
                    )
                total_open_orders += len(open_orders)
            else:
                client_section.append("  No open orders")
            client_section.append("\nREJECTED ORDERS:")
            if rejected_orders:
                for reject in rejected_orders:
                    try:
                        ts = datetime.strptime(reject['timestamp'], '%d-%m-%Y %H:%M:%S')
                        formatted_ts = ts.strftime('%Y-%m-%d %H:%M:%S')
                    except:
                        formatted_ts = reject['timestamp']
                    client_section.append(
                        f"  {formatted_ts} | ID: {reject['order_id']:<12} | {reject['symbol']:<15} | Reason: {reject['reason'][:55]}..."
                    )
                total_rejections += len(rejected_orders)
            else:
                client_section.append("  No rejected orders")
            report_content.append('\n'.join(client_section))
            total_clients += 1
        except Exception as e:
            error_msg = f"\nError processing {client_id}: {str(e)}"
            report_content.append(error_msg)
    report_content.append("\n" + "=" * 60)
    report_content.append("SUMMARY STATISTICS")
    report_content.append("-" * 60)
    report_content.append(f"Total Clients Processed: {total_clients}")
    report_content.append(f"Total Positions Found:   {total_positions}")
    report_content.append(f"Total Open Orders:       {total_open_orders}")
    report_content.append(f"Total Rejected Orders:   {total_rejections}")
    report_content.append("=" * 60)
    with open(filename, 'w') as f:
        f.write('\n'.join(report_content))
    print(f"Report generated: {filename}")
    return filename

# ------------------------------
# Sell Order Position Monitor (logs to file)
# ------------------------------
def position_monitor_sell(stop_event):
    """Monitor and log sell positions and orders to client-specific log files."""
    while not stop_event.is_set():
        for client_id in client_details:
            header_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            log_entry = f"\n[{header_time}] {client_id} SELL MONITOR\n" + "-" * 60 + "\n"
            try:
                api = client_apis.get(client_id)
                if not api:
                    continue
                # Log current holdings (with stock names)
                log_entry += "CURRENT HOLDINGS (with Stock Names):\n"
                holding_response = api.get_holding(clientID=client_id)
                if holding_response.get('type') == 'success':
                    holdings = holding_response['result']['RMSHoldings']['Holdings']
                    if holdings:
                        for isin, h in holdings.items():
                            instrument_id = h.get('ExchangeNSEInstrumentID') or h.get('ExchangeNSEInstrumentId')
                            description = h.get('Description', 'Unknown')
                            quantity = h.get('HoldingQuantity', '0')
                            if instrument_id:
                                try:
                                    ex_id_int = int(instrument_id)
                                    stock_info = stock_details.get(ex_id_int, {})
                                    stock_name = stock_info.get('name', description)
                                except Exception:
                                    stock_name = description
                            else:
                                stock_name = description
                            log_entry += f"  {stock_name} (ID: {instrument_id}): {quantity}\n"
                    else:
                        log_entry += "  No holdings available\n"
                else:
                    log_entry += "  Error fetching holdings\n"
                # Log live positions:
                log_entry += "\nLIVE POSITIONS:\n"
                pos_response = api.get_position_daywise(clientID=client_id)
                if pos_response.get('type') == 'success':
                    positions = pos_response['result'].get('positionList', [])
                    if positions:
                        for pos in positions:
                            trading_symbol = pos.get('TradingSymbol', 'N/A')
                            quantity = pos.get('Quantity', 'N/A')
                            avg_price = pos.get('AveragePrice', 'N/A')
                            log_entry += f"  {trading_symbol}: {quantity} shares, Avg Price: {avg_price}\n"
                    else:
                        log_entry += "  No live positions available\n"
                else:
                    log_entry += "  Error fetching live positions\n"
                # Log open sell orders:
                open_orders, rejected_orders = fetch_client_orders(api, client_id)
                log_entry += "\nOPEN SELL ORDERS:\n"
                if open_orders:
                    for o in open_orders:
                        order_time = o.get('OrderGeneratedDateTime', header_time)
                        trading_symbol = o.get('TradingSymbol', 'N/A')
                        quantity = o.get('OrderQuantity', 'N/A')
                        price = o.get('OrderPrice', 'N/A')
                        status = o.get('OrderStatus', 'N/A')
                        log_entry += f"  {order_time} - {trading_symbol} : {quantity} shares @ {price} ({status})\n"
                else:
                    log_entry += "  No open sell orders\n"
                # Log rejected orders:
                log_entry += "\nREJECTED ORDERS:\n"
                if rejected_orders:
                    for r in rejected_orders:
                        raw_ts = r.get('timestamp', header_time)
                        try:
                            dt = datetime.strptime(raw_ts, '%d-%m-%Y %H:%M:%S')
                            formatted_ts = dt.strftime('%Y-%m-%d %H:%M:%S')
                        except Exception:
                            formatted_ts = raw_ts
                        symbol = r.get('symbol', 'N/A')
                        reason = r.get('reason', 'No reason provided')
                        log_entry += f"  {formatted_ts} - {symbol}: {reason}\n"
                else:
                    log_entry += "  No rejected orders\n"
                with open(f"sell_monitor_{client_id}.log", "a") as f:
                    f.write(log_entry + "\n")
            except Exception as e:
                error_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                error_msg = f"[{error_time}] Monitoring error for {client_id}: {str(e)}"
                with open(f"sell_monitor_{client_id}.log", "a") as f:
                    f.write(error_msg + "\n")
        time.sleep(10)

# ------------------------------
# Final Summary Report (Validation)
# ------------------------------
def parse_order_time(ts_str):
    """Parse order timestamp using multiple potential formats."""
    for fmt in (
        '%Y-%m-%d %H:%M:%S.%f',  # With microseconds
        '%Y-%m-%d %H:%M:%S',     # Without microseconds
        '%d-%m-%Y %H:%M:%S',     # Day-month-year format
        '%m/%d/%Y %H:%M:%S',     # Month/day/year (common in some APIs)
        '%d-%b-%Y %H:%M:%S',     # With abbreviated month (e.g., 13-Feb-2024)
    ):
        try:
            return datetime.strptime(ts_str, fmt)
        except ValueError:
            continue
    print(f"Warning: Failed to parse timestamp '{ts_str}'")
    return None  # Explicitly return None if parsing fails

def display_final_summary_sell(client_details, order_responses, order_start_time):
    """
    Display summary report including orders placed after order_start_time and live positions.
    Only orders with status exactly 'FILLED' are counted as executed stock.
    """
    print("\n==================== Final Summary Report ====================")
    for client_id, params in client_details.items():
        print(f"\nClient: {client_id}")
        print("-" * 70)
        header = f"{'Instrument':<15} | {'Target':>6} | {'Executed':>8} | {'Open':>6} | {'Rejected':>8} | {'Live Pos':>8}"
        print(header)
        print("-" * len(header))
        api = client_apis.get(client_id)
        if not api:
            continue

        # Fetch the latest order book and live positions
        try:
            order_book_response = api.get_order_book(clientID=client_id)
            pos_response = api.get_position_daywise(clientID=client_id)
        except Exception as e:
            print(f"Error fetching data for {client_id}: {str(e)}")
            continue

        # Process live positions
        live_positions = {}
        if pos_response.get('type') == 'success':
            positions_list = pos_response['result'].get('positionList', [])
            if positions_list:
                for pos in positions_list:
                    trading_symbol = pos.get('TradingSymbol', 'N/A')
                    quantity = pos.get('Quantity', 0)
                    # Try to get the instrument ID using one of the available keys
                    inst_id = pos.get('ExchangeInstrumentID') or pos.get('ExchangeNSEInstrumentID') or pos.get('ExchangeNSEInstrumentId')
                    if inst_id:
                        try:
                            inst_id = int(inst_id)
                            live_positions[inst_id] = {
                                'trading_symbol': trading_symbol,
                                'quantity': quantity
                            }
                        except Exception as e:
                            print(f"Error processing live position for {client_id}: {e}")
                            continue

        # Process orders placed after order_start_time
        instrument_orders = defaultdict(lambda: {'executed': 0, 'open': 0, 'rejected': 0})
        orders = order_book_response.get('result', []) if order_book_response.get('type') == 'success' else []
        for order in orders:
            try:
                # Get instrument ID from multiple potential keys
                order_inst_id = order.get('ExchangeInstrumentID') or order.get('ExchangeNSEInstrumentID') or order.get('ExchangeNSEInstrumentId')
                if not order_inst_id:
                    continue
                order_inst_id = int(order_inst_id)
                status = order.get('OrderStatus', '').upper()
                order_time = parse_order_time(order.get('OrderGeneratedDateTime', ''))
                # Skip orders placed before the strategy started
                if not order_time or order_time < order_start_time:
                    continue
                order_qty = int(order.get('OrderQuantity', 0))
                # Only count orders that are exactly FILLED as executed stock.
                if status == "FILLED":
                    instrument_orders[order_inst_id]['executed'] += order_qty
                elif status in ["NEW", "REPLACED", "OPEN", "PARTIALLY_FILLED"]:
                    instrument_orders[order_inst_id]['open'] += order_qty
                elif status in ["REJECTED","CANCELLED"]:
                    instrument_orders[order_inst_id]['rejected'] += order_qty
            except Exception as e:
                print(f"Error processing order for {client_id}: {str(e)}")

        # Display summary for each instrument defined in the client's sell details
        for ex_id, target_qty in params.get('total_quantity_to_sell', {}).items():
            ex_id_int = int(ex_id)
            stock_name = stock_details.get(ex_id_int, {}).get('name', f"ID:{ex_id}")
            orders_data = instrument_orders.get(ex_id_int, {'executed': 0, 'open': 0, 'rejected': 0})
            live_qty = live_positions.get(ex_id_int, {}).get('quantity', 0)
            print(f"{stock_name:<15} | {target_qty:6} | {orders_data['executed']:8} | {orders_data['open']:6} | {orders_data['rejected']:8} | {live_qty:8}")

        # Optionally, print detailed live positions information
        print("\nDetailed Live Positions:")
        if live_positions:
            for inst_id, info in live_positions.items():
                print(f"  {info['trading_symbol']}: {info['quantity']} shares")
        else:
            print("  No live positions available.")

# ------------------------------
# Configuration: Stock & Client Details
# ------------------------------
# These details should be defined externally.
stock_details = stock_details  # Replace with your actual stock details dictionary
client_details = client_sell_details  # Replace with your actual client sell details dictionary
order_start_time = datetime.now()

# ------------------------------
# Main Execution
# ------------------------------
if __name__ == "__main__":
    # Start the monitoring thread (logs order and holding info to files)
    stop_event = Event()
    monitor_thread = Thread(target=position_monitor_sell, args=(stop_event,))
    monitor_thread.daemon = True
    monitor_thread.start()

    # Record the actual start time before placing orders
    order_start_time = order_start_time

    # Execute sell strategy (total_time in seconds, e.g., 60 seconds)
    order_responses = execute_sell_strategy(stock_details, client_details, total_time=3*60)

    # Modify orders periodically
    for _ in range(20):
        modify_sell_orders(order_responses)
        time.sleep(30)

    # Stop monitoring
    stop_event.set()
    monitor_thread.join(timeout=5)

    # Generate final summary with correct start time and live positions.
    # Only orders with status 'FILLED' (placed after order_start_time) are added as executed stock.
    display_final_summary_sell(client_details, order_responses, order_start_time)

    # Generate text report if needed
    #generate_text_summary(client_details)
```

```python
import time
import json
import os
import threading
from threading import Event, Thread
from datetime import datetime
from collections import defaultdict
import tkinter as tk
from tkinter import ttk
import queue

# ------------------------------
# Market Data API & Client APIs
# ------------------------------
# (Initialize your API instances as needed)
xtm = xtm  # Market Data API instance
client_apis = {
    "ALG347": xti,
    "ADV21": xti2,
    "ADV26": xti3,
    "MRG1": xti5,
    "SBJ01": xti6,
    "SBJ20": xti7,
    "SBJ22": xti8,
    'ALG303': xti9,
    'ALG337': xti10,
    'ALG345': xti11,
    'ALG328': xti12,
    'SBJ15': xti13,
    'ADV03': xti14,
    'ADV27': xti16,
    'ALG301': xti17,
    'SBJ04': xti18,
    'ALG300': xti19,
    # Add other clients as needed
}

# ------------------------------
# Global variables for instrument locks and LTP cache
# ------------------------------
instrument_locks = defaultdict(threading.Lock)
ltp_cache = {}

def fetch_current_ltp(exchange_instrument_id):
    """Fetch LTP with price adjustment logic based on high/low thresholds."""
    cache_expiry = 2  # seconds
    with instrument_locks[exchange_instrument_id]:
        now = time.time()
        cached = ltp_cache.get(exchange_instrument_id)
        if cached and (now - cached['timestamp']) < cache_expiry:
            return cached['ltp']
        try:
            instruments = [{
                'exchangeSegment': 1,
                'exchangeInstrumentID': int(exchange_instrument_id)
            }]
            unsub_response = xtm.send_unsubscription(Instruments=instruments, xtsMessageCode=1502)
            sub_response = xtm.send_subscription(Instruments=instruments, xtsMessageCode=1502)
            if not sub_response.get('result'):
                return None

            original_ltp = None
            for quote_json in sub_response['result']['listQuotes']:
                quote_data = json.loads(quote_json)
                original_ltp = quote_data['Touchline']['LastTradedPrice']
                break

            if original_ltp is None:
                return None

            ex_id_int = int(exchange_instrument_id)
            stock_info = stock_details.get(ex_id_int, {})
            tick_size = stock_info.get('tick_size', 0.05)
            high = float(stock_info.get('high_string', 0))
            low = float(stock_info.get('low_string', 0))

            adjusted_ltp = original_ltp
            adjusted_ltp = round(adjusted_ltp / tick_size) * tick_size

            if high > 0 and low > 0:
                if adjusted_ltp >= high or adjusted_ltp <= low:
                    final_ltp = round(original_ltp / tick_size) * tick_size
                else:
                    final_ltp = adjusted_ltp
            else:
                final_ltp = adjusted_ltp

            ltp_cache[exchange_instrument_id] = {
                'ltp': final_ltp,
                'timestamp': time.time()
            }
            return final_ltp
        except Exception as e:
            print(f"LTP Error: {str(e)}")
            return None

def fetch_sell_ltp(exchange_instrument_id):
    """Fetch LTP for sell orders with tick rounding."""
    base_ltp = fetch_current_ltp(exchange_instrument_id)
    if base_ltp is None:
        return None
    ex_id_int = int(exchange_instrument_id)
    tick_size = stock_details.get(ex_id_int, {}).get('tick_size', 0.05)
    adjusted_ltp = round(base_ltp / tick_size) * tick_size
    return adjusted_ltp

# ------------------------------
# Holding and Order Functions
# ------------------------------
def get_holding_quantity(client_id, exchange_instrument_id, api):
    """Fetch holding quantity for a specific instrument and client."""
    try:
        response = api.get_holding(clientID=client_id)
        if response['type'] != 'success':
            print(f"Error fetching holdings for {client_id}: {response.get('message')}")
            return None
        holdings = response['result'].get('RMSHoldings', {}).get('Holdings', {})
        for isin, details in holdings.items():
            if details.get('ExchangeNSEInstrumentId') == exchange_instrument_id:
                return int(details.get('HoldingQuantity', 0))
        return 0
    except Exception as e:
        print(f"Exception in get_holding_quantity for {client_id}: {e}")
        return None

def validate_holding_quantity(client_id, exchange_instrument_id, total_quantity_to_sell, api):
    """Validate if client has enough holdings to sell considering open orders."""
    holding_qty = get_holding_quantity(client_id, exchange_instrument_id, api)
    if holding_qty is None:
        print(f"Validation failed: Could not fetch holdings for {client_id}")
        return False

    open_sell_qty = 0
    try:
        order_response = api.get_order_book(clientID=client_id)
        if order_response.get('type') == 'success':
            for order in order_response['result']:
                order_instrument_id = str(order.get('ExchangeInstrumentID', ''))
                target_instrument_id = str(exchange_instrument_id)
                if order_instrument_id == target_instrument_id:
                    if order.get('OrderSide', '').upper() == 'SELL':
                        status = order.get('OrderStatus', '').upper()
                        if status in ['NEW', 'REPLACED', 'PARTIALLY_FILLED', 'OPEN']:
                            open_sell_qty += order.get('OrderQuantity', 0)
    except Exception as e:
        print(f"Error fetching open orders for {client_id}: {str(e)}")
        return False

    available_qty = holding_qty - open_sell_qty
    if available_qty < total_quantity_to_sell:
        print(f"Validation failed: {client_id} has {holding_qty} units, {open_sell_qty} in open orders. Available: {available_qty}, needed: {total_quantity_to_sell}")
        return False
    return True

def place_sell_order(client_id, exchange_instrument_id, ltp, quantity, order_responses, lock):
    """Place sell order using client-specific API."""
    try:
        api = client_apis.get(client_id)
        if api is None:
            print(f"No API found for {client_id}")
            return 0
        response = api.place_order(
            exchangeSegment=api.EXCHANGE_NSECM,
            exchangeInstrumentID=exchange_instrument_id,
            productType=api.PRODUCT_CNC,
            orderType=api.ORDER_TYPE_LIMIT,
            orderSide=api.TRANSACTION_TYPE_SELL,
            timeInForce=api.VALIDITY_DAY,
            disclosedQuantity=0,
            orderQuantity=quantity,
            limitPrice=ltp,
            stopPrice=0,
            orderUniqueIdentifier=str(datetime.now().timestamp()),
            clientID=client_id
        )
        if response['type'] != 'error':
            order_id = response['result']['AppOrderID']
            with lock:
                order_responses.setdefault((exchange_instrument_id, client_id), []).append({
                    'order_id': order_id,
                    'quantity': quantity,
                    'client_id': client_id
                })
            print(f"[{datetime.now()}] {client_id} Sell Order {order_id}: {quantity} @ {ltp}")
            return quantity
        print(f"Sell order failed {client_id}: {response.get('message')}")
        return 0
    except Exception as e:
        print(f"Sell Order Error {client_id}: {str(e)}")
        return 0

def schedule_sell_orders(client_id, exchange_instrument_id, details, params, total_time, order_responses, lock):
    """Schedule sell orders for a client and instrument."""
    api = client_apis.get(client_id)
    if not api:
        return

    total_quantity = params['total_quantity_to_sell'].get(exchange_instrument_id, 0)
    if total_quantity <= 0:
        return

    if not validate_holding_quantity(client_id, int(exchange_instrument_id), total_quantity, api):
        return

    ltp = fetch_sell_ltp(exchange_instrument_id)
    if not ltp:
        return

    try:
        single_order_qty = max(1, int(params['single_order_value'] / ltp))
        n_orders = max(1, (total_quantity + single_order_qty - 1) // single_order_qty)
        interval = total_time / n_orders
    except Exception as e:
        print(f"Calculation Error {client_id}: {str(e)}")
        return

    print(f"\n{client_id} Selling {details['name']}:")
    print(f"Total: {total_quantity}, Orders: {n_orders}, Interval: {interval:.2f}s")

    remaining = total_quantity
    start_time = time.time()
    for _ in range(n_orders):
        if remaining <= 0:
            break
        current_ltp = fetch_sell_ltp(exchange_instrument_id)
        if not current_ltp:
            continue
        qty = min(single_order_qty, remaining)
        executed = place_sell_order(client_id, exchange_instrument_id, current_ltp, qty, order_responses, lock)
        remaining -= executed
        elapsed = time.time() - start_time
        sleep_time = max(0, interval - (elapsed % interval))
        time.sleep(sleep_time)

def check_order_status(api, client_id, order_id):
    """Check order status from the order book using the order ID."""
    try:
        response = api.get_order_book(clientID=client_id)
        if response.get('type') == 'success':
            for order in response['result']:
                if order.get('AppOrderID') == order_id:
                    status = order.get('OrderStatus', '').upper()
                    return status in ['NEW', 'REPLACED', 'PARTIALLY_FILLED', 'OPEN']
        return False
    except Exception as e:
        print(f"Status check failed for {order_id}: {str(e)}")
        return False

def modify_sell_orders(order_responses):
    """Modify live orders by checking their status and updating the limit price if needed."""
    overall_modified = False
    for (ex_id, client_id), orders in order_responses.items():
        api = client_apis.get(client_id)
        if not api:
            continue
        modified_for_pair = False
        for order_info in orders:
            if check_order_status(api, client_id, order_info['order_id']):
                current_ltp = fetch_sell_ltp(ex_id)
                if current_ltp:
                    response = api.modify_order(
                        appOrderID=order_info['order_id'],
                        modifiedProductType=api.PRODUCT_CNC,
                        modifiedOrderType=api.ORDER_TYPE_LIMIT,
                        modifiedOrderQuantity=order_info['quantity'],
                        modifiedDisclosedQuantity=0,
                        modifiedLimitPrice=current_ltp,
                        modifiedStopPrice=0,
                        modifiedTimeInForce=api.VALIDITY_DAY,
                        orderUniqueIdentifier=str(datetime.now().timestamp()),
                        clientID=client_id
                    )
                    print(f"{client_id} Modified sell order {order_info['order_id']} @ {current_ltp}")
                    overall_modified = True
                    modified_for_pair = True
        if not modified_for_pair:
            print(f"{client_id} - For instrument {ex_id}: No order Left to modify based on OrderID.")
    if not overall_modified:
        print("No order Left to modify based on OrderID.")

# ------------------------------
# Strategy Execution
# ------------------------------
def execute_sell_strategy(stock_details, client_details, total_time):
    """Orchestrate sell order placement across clients and stocks."""
    order_responses = {}
    lock = threading.Lock()
    threads = []
    for client_id, params in client_details.items():
        for ex_id, details in stock_details.items():
            if ex_id not in params.get('total_quantity_to_sell', {}):
                continue
            thread = threading.Thread(
                target=schedule_sell_orders,
                args=(client_id, ex_id, details, params, total_time, order_responses, lock)
            )
            threads.append(thread)
            thread.start()
    for thread in threads:
        thread.join()
    return order_responses

# ------------------------------
# Order Fetching and Reporting Helpers
# ------------------------------
def fetch_client_orders(api, client_id):
    """Fetch all orders for a given client via the API."""
    try:
        order_response = api.get_order_book(clientID=client_id)
        if order_response.get('type') != 'success':
            return [], []
        orders = order_response.get('result', [])
        open_orders = []
        rejected_orders = []
        for order in orders:
            if order['OrderStatus'] in ['New', 'Replaced']:
                open_orders.append(order)
            elif order['OrderStatus'] == 'Rejected':
                rejected_orders.append({
                    'symbol': order['TradingSymbol'],
                    'order_id': order.get('AppOrderID', 'N/A'),
                    'timestamp': order.get('OrderGeneratedDateTime', 'N/A'),
                    'reason': order.get('CancelRejectReason', 'No reason provided')
                })
        return open_orders, rejected_orders
    except Exception as e:
        print(f"Order fetch error for {client_id}: {str(e)}")
        return [], []

def generate_text_summary(client_details):
    """Generate a comprehensive text report with positions, orders, and rejections."""
    if not os.path.exists('reports'):
        os.makedirs('reports')
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"reports/trading_summary_{timestamp}.txt"
    report_content = []
    report_content.append("=" * 60)
    report_content.append(f"TRADING SUMMARY REPORT - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    report_content.append("=" * 60 + "\n")
    total_clients = 0
    total_positions = 0
    total_open_orders = 0
    total_rejections = 0
    for client_id in client_details:
        try:
            api = client_apis.get(client_id)
            if not api:
                continue
            client_section = []
            client_section.append(f"\nCLIENT: {client_id}")
            client_section.append("-" * 60)
            positions = []
            pos_response = api.get_position_daywise(clientID=client_id)
            if pos_response.get('type') == 'success':
                positions = pos_response['result'].get('positionList', [])
            client_section.append("\nCURRENT POSITIONS:")
            if positions:
                for pos in positions:
                    client_section.append(
                        f"  {pos['TradingSymbol']:<15} | Quantity: {pos['Quantity']:<6} | Avg Price: {pos.get('AveragePrice', 'N/A')}"
                    )
                total_positions += len(positions)
            else:
                client_section.append("  No positions found")
            open_orders, rejected_orders = fetch_client_orders(api, client_id)
            client_section.append("\nOPEN ORDERS:")
            if open_orders:
                for order in open_orders:
                    client_section.append(
                        f"  {order['TradingSymbol']:<15} | ID: {order['AppOrderID']:<12} | Qty: {order['OrderQuantity']:<5} | Price: {order['OrderPrice']:<8} | Status: {order['OrderStatus']}"
                    )
                total_open_orders += len(open_orders)
            else:
                client_section.append("  No open orders")
            client_section.append("\nREJECTED ORDERS:")
            if rejected_orders:
                for reject in rejected_orders:
                    try:
                        ts = datetime.strptime(reject['timestamp'], '%d-%m-%Y %H:%M:%S')
                        formatted_ts = ts.strftime('%Y-%m-%d %H:%M:%S')
                    except:
                        formatted_ts = reject['timestamp']
                    client_section.append(
                        f"  {formatted_ts} | ID: {reject['order_id']:<12} | {reject['symbol']:<15} | Reason: {reject['reason'][:55]}..."
                    )
                total_rejections += len(rejected_orders)
            else:
                client_section.append("  No rejected orders")
            report_content.append('\n'.join(client_section))
            total_clients += 1
        except Exception as e:
            error_msg = f"\nError processing {client_id}: {str(e)}"
            report_content.append(error_msg)
    report_content.append("\n" + "=" * 60)
    report_content.append("SUMMARY STATISTICS")
    report_content.append("-" * 60)
    report_content.append(f"Total Clients Processed: {total_clients}")
    report_content.append(f"Total Positions Found:   {total_positions}")
    report_content.append(f"Total Open Orders:       {total_open_orders}")
    report_content.append(f"Total Rejected Orders:   {total_rejections}")
    report_content.append("=" * 60)
    with open(filename, 'w') as f:
        f.write('\n'.join(report_content))
    print(f"Report generated: {filename}")
    return filename

# ------------------------------
# Sell Order Position Monitor (sends to GUI)
# ------------------------------
def position_monitor_sell(stop_event, data_queue):
    """Monitor and send sell positions and orders to GUI via queue."""
    while not stop_event.is_set():
        data = {}
        for client_id in client_details:
            header_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            log_entry = f"[{header_time}] {client_id} SELL MONITOR\n" + "-" * 60 + "\n"
            try:
                api = client_apis.get(client_id)
                if not api:
                    continue
                log_entry += "CURRENT HOLDINGS (with Stock Names):\n"
                holding_response = api.get_holding(clientID=client_id)
                if holding_response.get('type') == 'success':
                    holdings = holding_response['result']['RMSHoldings']['Holdings']
                    if holdings:
                        for isin, h in holdings.items():
                            instrument_id = h.get('ExchangeNSEInstrumentID') or h.get('ExchangeNSEInstrumentId')
                            description = h.get('Description', 'Unknown')
                            quantity = h.get('HoldingQuantity', '0')
                            if instrument_id:
                                try:
                                    ex_id_int = int(instrument_id)
                                    stock_info = stock_details.get(ex_id_int, {})
                                    stock_name = stock_info.get('name', description)
                                except Exception:
                                    stock_name = description
                            else:
                                stock_name = description
                            log_entry += f"  {stock_name} (ID: {instrument_id}): {quantity}\n"
                    else:
                        log_entry += "  No holdings available\n"
                else:
                    log_entry += "  Error fetching holdings\n"
                log_entry += "\nLIVE POSITIONS:\n"
                pos_response = api.get_position_daywise(clientID=client_id)
                if pos_response.get('type') == 'success':
                    positions = pos_response['result'].get('positionList', [])
                    if positions:
                        for pos in positions:
                            trading_symbol = pos.get('TradingSymbol', 'N/A')
                            quantity = pos.get('Quantity', 'N/A')
                            avg_price = pos.get('AveragePrice', 'N/A')
                            log_entry += f"  {trading_symbol}: {quantity} shares, Avg Price: {avg_price}\n"
                    else:
                        log_entry += "  No live positions available\n"
                else:
                    log_entry += "  Error fetching live positions\n"
                open_orders, rejected_orders = fetch_client_orders(api, client_id)
                log_entry += "\nOPEN SELL ORDERS:\n"
                if open_orders:
                    for o in open_orders:
                        order_time = o.get('OrderGeneratedDateTime', header_time)
                        trading_symbol = o.get('TradingSymbol', 'N/A')
                        quantity = o.get('OrderQuantity', 'N/A')
                        price = o.get('OrderPrice', 'N/A')
                        status = o.get('OrderStatus', 'N/A')
                        log_entry += f"  {order_time} - {trading_symbol} : {quantity} shares @ {price} ({status})\n"
                else:
                    log_entry += "  No open sell orders\n"
                log_entry += "\nREJECTED ORDERS:\n"
                if rejected_orders:
                    for r in rejected_orders:
                        raw_ts = r.get('timestamp', header_time)
                        try:
                            dt = datetime.strptime(raw_ts, '%d-%m-%Y %H:%M:%S')
                            formatted_ts = dt.strftime('%Y-%m-%d %H:%M:%S')
                        except Exception:
                            formatted_ts = raw_ts
                        symbol = r.get('symbol', 'N/A')
                        reason = r.get('reason', 'No reason provided')
                        log_entry += f"  {formatted_ts} - {symbol}: {reason}\n"
                else:
                    log_entry += "  No rejected orders\n"
                data[client_id] = log_entry
            except Exception as e:
                error_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                error_msg = f"[{error_time}] Monitoring error for {client_id}: {str(e)}"
                data[client_id] = error_msg
        data_queue.put(data)
        time.sleep(10)

# ------------------------------
# Final Summary Report (Validation)
# ------------------------------
def parse_order_time(ts_str):
    """Parse order timestamp using multiple potential formats."""
    for fmt in (
        '%Y-%m-%d %H:%M:%S.%f',
        '%Y-%m-%d %H:%M:%S',
        '%d-%m-%Y %H:%M:%S',
        '%m/%d/%Y %H:%M:%S',
        '%d-%b-%Y %H:%M:%S',
    ):
        try:
            return datetime.strptime(ts_str, fmt)
        except ValueError:
            continue
    print(f"Warning: Failed to parse timestamp '{ts_str}'")
    return None

def display_final_summary_sell(client_details, order_responses, order_start_time):
    """Display summary report including orders placed after order_start_time and live positions."""
    print("\n==================== Final Summary Report ====================")
    for client_id, params in client_details.items():
        print(f"\nClient: {client_id}")
        print("-" * 70)
        header = f"{'Instrument':<15} | {'Target':>6} | {'Executed':>8} | {'Open':>6} | {'Rejected':>8} | {'Live Pos':>8}"
        print(header)
        print("-" * len(header))
        api = client_apis.get(client_id)
        if not api:
            continue
        try:
            order_book_response = api.get_order_book(clientID=client_id)
            pos_response = api.get_position_daywise(clientID=client_id)
        except Exception as e:
            print(f"Error fetching data for {client_id}: {str(e)}")
            continue
        live_positions = {}
        if pos_response.get('type') == 'success':
            positions_list = pos_response['result'].get('positionList', [])
            if positions_list:
                for pos in positions_list:
                    trading_symbol = pos.get('TradingSymbol', 'N/A')
                    quantity = pos.get('Quantity', 0)
                    inst_id = pos.get('ExchangeInstrumentID') or pos.get('ExchangeNSEInstrumentID') or pos.get('ExchangeNSEInstrumentId')
                    if inst_id:
                        try:
                            inst_id = int(inst_id)
                            live_positions[inst_id] = {
                                'trading_symbol': trading_symbol,
                                'quantity': quantity
                            }
                        except Exception as e:
                            print(f"Error processing live position for {client_id}: {e}")
                            continue
        instrument_orders = defaultdict(lambda: {'executed': 0, 'open': 0, 'rejected': 0})
        orders = order_book_response.get('result', []) if order_book_response.get('type') == 'success' else []
        for order in orders:
            try:
                order_inst_id = order.get('ExchangeInstrumentID') or order.get('ExchangeNSEInstrumentID') or order.get('ExchangeNSEInstrumentId')
                if not order_inst_id:
                    continue
                order_inst_id = int(order_inst_id)
                status = order.get('OrderStatus', '').upper()
                order_time = parse_order_time(order.get('OrderGeneratedDateTime', ''))
                if not order_time or order_time < order_start_time:
                    continue
                order_qty = int(order.get('OrderQuantity', 0))
                if status == "FILLED":
                    instrument_orders[order_inst_id]['executed'] += order_qty
                elif status in ["NEW", "REPLACED", "OPEN", "PARTIALLY_FILLED"]:
                    instrument_orders[order_inst_id]['open'] += order_qty
                elif status in ["REJECTED", "CANCELLED"]:
                    instrument_orders[order_inst_id]['rejected'] += order_qty
            except Exception as e:
                print(f"Error processing order for {client_id}: {str(e)}")
        for ex_id, target_qty in params.get('total_quantity_to_sell', {}).items():
            ex_id_int = int(ex_id)
            stock_name = stock_details.get(ex_id_int, {}).get('name', f"ID:{ex_id}")
            orders_data = instrument_orders.get(ex_id_int, {'executed': 0, 'open': 0, 'rejected': 0})
            live_qty = live_positions.get(ex_id_int, {}).get('quantity', 0)
            print(f"{stock_name:<15} | {target_qty:6} | {orders_data['executed']:8} | {orders_data['open']:6} | {orders_data['rejected']:8} | {live_qty:8}")
        print("\nDetailed Live Positions:")
        if live_positions:
            for inst_id, info in live_positions.items():
                print(f"  {info['trading_symbol']}: {info['quantity']} shares")
        else:
            print("  No live positions available.")

# ------------------------------
# Configuration: Stock & Client Details
# ------------------------------
stock_details = stock_details  # Replace with your actual stock details dictionary
client_details = client_sell_details  # Replace with your actual client sell details dictionary
order_start_time = datetime.now()

# ------------------------------
# GUI Setup
# ------------------------------
root = tk.Tk()
root.title("Sell Monitor")
notebook = ttk.Notebook(root)
notebook.pack(fill='both', expand=True)

client_frames = {}
for client_id in client_details:
    frame = ttk.Frame(notebook)
    notebook.add(frame, text=client_id)
    text_widget = tk.Text(frame, wrap='word', height=20, width=80)
    scrollbar = tk.Scrollbar(frame, command=text_widget.yview)
    text_widget.config(yscrollcommand=scrollbar.set)
    text_widget.pack(side='left', fill='both', expand=True)
    scrollbar.pack(side='right', fill='y')
    client_frames[client_id] = text_widget

data_queue = queue.Queue()

def update_gui():
    """Update GUI with data from the queue."""
    try:
        while True:
            data = data_queue.get_nowait()
            for client_id, text in data.items():
                text_widget = client_frames[client_id]
                text_widget.delete('1.0', tk.END)
                text_widget.insert(tk.END, text)
    except queue.Empty:
        pass
    root.after(100, update_gui)

update_gui()

# ------------------------------
# Main Execution
# ------------------------------
if __name__ == "__main__":
    stop_event = Event()
    monitor_thread = Thread(target=position_monitor_sell, args=(stop_event, data_queue))
    monitor_thread.daemon = True
    monitor_thread.start()

    order_start_time = datetime.now()

    def run_trading():
        order_responses = execute_sell_strategy(stock_details, client_details, total_time=3*60)
        for _ in range(20):
            modify_sell_orders(order_responses)
            time.sleep(30)
        display_final_summary_sell(client_details, order_responses, order_start_time)

    trading_thread = Thread(target=run_trading)
    trading_thread.start()

    def on_closing():
        stop_event.set()
        root.destroy()

    root.protocol("WM_DELETE_WINDOW", on_closing)
    root.mainloop()
```

```python
import pandas as pd
from collections import defaultdict
from datetime import datetime

# Assuming client_apis is defined as in your script
# client_apis = {"ALG347": xti, ...}

def parse_order_time(ts_str):
    """Parse order timestamp using multiple potential formats."""
    for fmt in (
        '%Y-%m-%d %H:%M:%S.%f',  # With microseconds
        '%Y-%m-%d %H:%M:%S',     # Without microseconds
        '%d-%m-%Y %H:%M:%S',     # Day-month-year format
        '%m/%d/%Y %H:%M:%S',     # Month/day/year
        '%d-%b-%Y %H:%M:%S',     # With abbreviated month (e.g., 13-Feb-2024)
    ):
        try:
            return datetime.strptime(ts_str, fmt)
        except ValueError:
            continue
    print(f"Warning: Failed to parse timestamp '{ts_str}'")
    return None

def generate_summary_data(client_details, order_start_time, stock_details):
    """
    Generate summary data for each client using live positions and order book details.
    
    Args:
        client_details (dict): Contains each client's sell configurations including target quantities.
        order_start_time (datetime): The time after which orders are considered.
        stock_details (dict): Global dictionary containing stock instrument details, keyed by instrument id.
    
    Returns:
        dict: Summary data with structure {client_id: [{"Instrument": str, "Target": int, ...}, ...]}
    """
    summary_data = {}
    
    for client_id, params in client_details.items():
        client_summary = []
        api = client_apis.get(client_id)
        if not api:
            print(f"API not found for client {client_id}")
            continue
        
        # Fetch latest data
        try:
            order_book_response = api.get_order_book(clientID=client_id)
            pos_response = api.get_position_daywise(clientID=client_id)
        except Exception as e:
            print(f"Error fetching data for {client_id}: {str(e)}")
            continue
        
        # Process live positions
        live_positions = {}
        if pos_response.get('type') == 'success':
            for pos in pos_response['result'].get('positionList', []):
                inst_id = (pos.get('ExchangeInstrumentID') or 
                          pos.get('ExchangeNSEInstrumentID') or 
                          pos.get('ExchangeNSEInstrumentId'))
                if inst_id:
                    try:
                        inst_id = int(inst_id)
                        live_positions[inst_id] = {
                            'trading_symbol': pos.get('TradingSymbol', 'N/A'),
                            'quantity': pos.get('Quantity', 0)
                        }
                    except Exception as e:
                        print(f"Error processing position for {client_id}: {e}")
        
        # Process orders
        instrument_orders = defaultdict(lambda: {'executed': 0, 'open': 0, 'rejected': 0})
        orders = order_book_response.get('result', []) if order_book_response.get('type') == 'success' else []
        
        for order in orders:
            order_inst_id = (order.get('ExchangeInstrumentID') or 
                            order.get('ExchangeNSEInstrumentID') or 
                            order.get('ExchangeNSEInstrumentId'))
            if not order_inst_id:
                continue
            try:
                order_inst_id = int(order_inst_id)
                status = order.get('OrderStatus', '').upper()
                order_time = parse_order_time(order.get('OrderGeneratedDateTime', ''))
                if not order_time or order_time < order_start_time:
                    continue
                order_qty = int(order.get('OrderQuantity', 0))
                if status == "FILLED":
                    instrument_orders[order_inst_id]['executed'] += order_qty
                elif status in ["NEW", "REPLACED", "OPEN", "PARTIALLY_FILLED"]:
                    instrument_orders[order_inst_id]['open'] += order_qty
                elif status == "REJECTED":
                    instrument_orders[order_inst_id]['rejected'] += order_qty
            except Exception as e:
                print(f"Error processing order for {client_id}: {str(e)}")
        
        # Build summary for each instrument
        for ex_id, target_qty in params.get('total_quantity_to_sell', {}).items():
            try:
                ex_id_int = int(ex_id)
            except Exception:
                continue
            stock_name = stock_details.get(ex_id_int, {}).get('name', f"ID:{ex_id}")
            orders_data = instrument_orders.get(ex_id_int, {'executed': 0, 'open': 0, 'rejected': 0})
            live_qty = live_positions.get(ex_id_int, {}).get('quantity', 0)
            
            client_summary.append({
                "Instrument": stock_name,
                "Target": target_qty,
                "Executed": orders_data['executed'],
                "Open": orders_data['open'],
                "Rejected": orders_data['rejected'],
                "Live Pos": live_qty
            })
        
        summary_data[client_id] = client_summary
    
    return summary_data

def export_summary_to_excel(summary_data, filename="trading_summary_sell.xlsx"):
    """
    Export summary data to an Excel file with one sheet per client.
    
    Args:
        summary_data (dict): Dictionary with client IDs as keys and lists of instrument details as values.
        filename (str): Name of the Excel file to create.
    """
    with pd.ExcelWriter(filename, engine='openpyxl') as writer:
        for client, instruments in summary_data.items():
            df = pd.DataFrame(instruments)
            desired_columns = ['Instrument', 'Target', 'Executed', 'Open', 'Rejected', 'Live Pos']
            df = df[[col for col in desired_columns if col in df.columns]]
            df.to_excel(writer, sheet_name=client, index=False)
    print(f"Excel report generated: {filename}")

def generate_excel_summary(client_details, order_start_time, stock_details, filename="trading_summary_sell.xlsx"):
    """Generate and export the Excel summary."""
    summary_data = generate_summary_data(client_details, order_start_time, stock_details)
    export_summary_to_excel(summary_data, filename)
generate_excel_summary(client_details, order_start_time, stock_details, filename="trading_summary_sell.xlsx")
```

```python
import pandas as pd
import time
import json
from threading import Lock

# Assuming xtm is a predefined API client
ltp_cache = {}
instrument_locks = {}  # Locks for each instrument to avoid race conditions

def get_instrument_details(symbol):
    """Get instrument details via API, defaulting TickSize to 0.05."""
    try:
        resp = xtm.get_equity_symbol(exchangeSegment=1, series='EQ', symbol=symbol)
        if resp.get('result'):
            inst = resp['result'][0]
            pb = inst.get('PriceBand', {})
            return {
                'ExchangeInstrumentID': inst['ExchangeInstrumentID'],
                'TickSize': inst.get('TickSize', 0.05),
                'HighString': pb.get('HighString', ''),
                'LowString': pb.get('LowString', '')
            }
    except Exception as e:
        print(f"Error fetching API for {symbol}: {e}")
    return None

def fetch_current_ltp(exch_id):
    """Fetch LTP, caching for 2s to avoid rapid recalls."""
    with instrument_locks.setdefault(exch_id, Lock()):
        now = time.time()
        cached = ltp_cache.get(exch_id)
        if cached and now - cached['timestamp'] < 2:
            return cached['ltp']
        try:
            instr = [{'exchangeSegment': 1, 'exchangeInstrumentID': int(exch_id)}]
            xtm.send_unsubscription(Instruments=instr, xtsMessageCode=1502)
            sub = xtm.send_subscription(Instruments=instr, xtsMessageCode=1502)
            for q in sub.get('result', {}).get('listQuotes', []):
                data = json.loads(q)
                ltp = data['Touchline']['LastTradedPrice']
                ltp_cache[exch_id] = {'ltp': ltp, 'timestamp': now}
                return ltp
        except Exception as e:
            print(f"LTP fetch error for {exch_id}: {e}")
    return None

def excel_to_trading_input(input_path):
    # --- Load master sheet ---
    master = pd.read_excel('master_instruments.xlsx', dtype=str)
    master['ISIN']    = master['ISIN'].str.strip()
    master['symbol1'] = master['symbol1'].str.strip().str.upper()
    master['instrumentID'] = master['instrumentID'].str.strip()
    # Drop duplicates so lookups are unique
    master = master.drop_duplicates(subset='ISIN', keep='first')
    master = master.drop_duplicates(subset='symbol1', keep='first')

    # For fast pairexistence check
    valid_pairs = set(zip(master['symbol1'], master['ISIN']))

    isin_map    = master.set_index('ISIN').to_dict('index')
    symbol1_map = master.set_index('symbol1').to_dict('index')

    # --- Load user orders ---
    df = pd.read_excel(input_path, sheet_name='Sheet1', dtype=str)
    df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)

    stock_details       = {}
    client_buy_details  = {}
    client_sell_details = {}
    cache               = {}  # symbol  resolved details or None

    def process(df_sub, is_buy):
        target = client_buy_details if is_buy else client_sell_details
        qty_col = 'QTY to Buy' if is_buy else 'QTY to Sell'

        for _, row in df_sub.iterrows():
            sym  = row['NSE Symbol'].strip().upper()
            isin = (row.get('ISIN') or '').strip().upper()

            # --- NEW: pair check against master ---
            if (sym, isin) not in valid_pairs:
                print(f"WARNING: Input pair NSE Symbol='{sym}' + ISIN='{isin}' not found in master_instruments.xlsx")
                continue

            # --- resolve details ---
            details = cache.get(sym)
            if details is None and sym not in cache:
                # 1) try API
                details = get_instrument_details(sym)
                # 2) fallback by ISIN
                if not details and isin in isin_map:
                    m = isin_map[isin]
                    details = {
                        'ExchangeInstrumentID': m['instrumentID'],
                        'TickSize': m.get('TickSize', 0.05),
                        'HighString': m.get('HighString', ''),
                        'LowString': m.get('LowString', '')
                    }
                # 3) fallback by symbol1
                if not details and sym in symbol1_map:
                    m = symbol1_map[sym]
                    details = {
                        'ExchangeInstrumentID': m['instrumentID'],
                        'TickSize': m.get('TickSize', 0.05),
                        'HighString': m.get('HighString', ''),
                        'LowString': m.get('LowString', '')
                    }
                cache[sym] = details  # may be None

            if not cache[sym]:
                # this should never happen now, since pair was valid
                continue

            exch_id = int(cache[sym]['ExchangeInstrumentID'])
            client  = row['USER ID']
            qty     = int(row[qty_col].replace(',', ''))
            sov     = int(row['SOV'].replace(',', ''))

            # 4) SOV vs LTP
            ltp = fetch_current_ltp(exch_id)
            if ltp is None or sov < ltp:
                print(f"Skipping {sym} (ID {exch_id}) for client {client}: SOV {sov} < LTP {ltp}")
                continue

            # 5) record stock_details
            if exch_id not in stock_details:
                stock_details[exch_id] = {
                    'name': sym,
                    'exchange_instrument_id': exch_id,
                    'tick_size': cache[sym]['TickSize'],
                    'high_string': cache[sym]['HighString'],
                    'low_string': cache[sym]['LowString']
                }

            # 6) record client orders
            if client not in target:
                target[client] = {
                    'single_order_value': sov,
                    f'total_quantity_to_{"buy" if is_buy else "sell"}': {}
                }
            key = f'total_quantity_to_{"buy" if is_buy else "sell"}'
            target[client][key][exch_id] = target[client][key].get(exch_id, 0) + qty

    process(df[df['QTY to Buy'].notna()], True)
    process(df[df['QTY to Sell'].notna()], False)

    # Filter out unused instruments
    used = set()
    for d in client_buy_details.values():  used |= set(d['total_quantity_to_buy'])
    for d in client_sell_details.values(): used |= set(d['total_quantity_to_sell'])
    stock_details = {i: v for i, v in stock_details.items() if i in used}

    return stock_details, client_buy_details, client_sell_details

if __name__ == "__main__":
    s, b, sl = excel_to_trading_input("Sample Excel Inputalg.xlsx")
    print("Instrument Details:", s)
    print("Buy Orders:", b)
    print("Sell Orders:", sl)
```

```python
import time
from datetime import datetime
import pandas as pd
import json
from threading import Lock

# --------------------------------------------------------
# 1) Load master_instruments.xlsx ONCE, globally
# --------------------------------------------------------
master_global = pd.read_excel('master_instruments.xlsx', dtype=str)

# Normalize & strip
master_global['ISIN']         = master_global['ISIN'].str.strip().str.upper()
master_global['symbol1']      = master_global['symbol1'].str.strip().str.upper()
master_global['instrumentID'] = master_global['instrumentID'].str.strip()
master_global['HighString']   = master_global['HighString'].astype(str).str.strip()
master_global['LowString']    = master_global['LowString'].astype(str).str.strip()
master_global['ticker']       = master_global['ticker'].astype(str).str.strip()

# Drop duplicates on (symbol1, ISIN) so merges are 1:1
master_global = master_global.drop_duplicates(subset=['symbol1', 'ISIN'], keep='first')

# Keep only the columns we need for direct lookups/merging
master_merge = master_global[['symbol1', 'ISIN', 'instrumentID', 'HighString', 'LowString', 'ticker']].copy()

# Create a mapping from TradingSymbol to instrumentID
symbol_to_instrument = master_global.set_index('symbol1')['instrumentID'].to_dict()

# --- Assumed Setup ---
# Define your client APIs and client details here.
client_apis = {
    "ALG347": xti,      # Replace with actual API instance for client ALG347
    #"Z46066": xti_zeb,  # Replace with actual API instance for client Z46066
    # "ADV21":  xti2,     # Replace with actual API instance for client ADV21
    # "ADV26":  xti3,     # Replace with actual API instance for client ADV26
    #"ALG330": xti4,
    # "MRG1": xti5,
    # #"SBJ01": xti6,
    # "SBJ20": xti7,
    # "SBJ22": xti8,
    # 'ALG303': xti9,
    # 'ALG337': xti10,
    # 'ALG345': xti11,
    # 'ALG328': xti12,
    # 'SBJ15': xti13,
    # 'ADV03': xti14,
    # 'ADV06': xti15,
    # 'ADV27': xti16,
    # 'ALG301': xti17,
    #'SBJ04': xti18,
    #'ALG300': xti19,
    # Replace with actual API instance for client ALG330
    # ... add all 25 clients as needed
}

# --- Core Functions ---

def get_client_live_positions(api, client_id):
    """
    Fetch live positions for a client and return a dictionary mapping instrumentID to quantity.
    """
    try:
        positions_response = api.get_position_daywise(clientID=client_id)
        if positions_response.get('type') == 'success':
            positions = positions_response['result'].get('positionList', [])
            positions_dict = {}
            for pos in positions:
                symbol = pos.get('TradingSymbol')
                qty = pos.get('Quantity', 0)
                instrument_id = symbol_to_instrument.get(symbol)
                if instrument_id:
                    positions_dict[int(instrument_id)] = int(qty)
            return positions_dict
        return {}
    except Exception as e:
        print(f"Error fetching positions for {client_id}: {e}")
        return {}

def get_client_orders(api, client_id):
    """
    Fetch live open and rejected orders for a client.
    Ensures function always returns exactly two lists.
    """
    try:
        order_response = api.get_order_book(clientID=client_id)
        if order_response.get('type') == 'success' and 'result' in order_response:
            orders = order_response['result'] or []
            open_orders = [order for order in orders if order.get('OrderStatus') in ['New', 'Replaced']]
            rejected_orders = [order for order in orders if order.get('OrderStatus') == 'Rejected']
            return open_orders, rejected_orders
        return [], []  # Ensure it always returns two lists
    except Exception as e:
        print(f"Error fetching orders for {client_id}: {e}")
        return [], []  # Default return values

def fetch_all_clients_data(client_details):
    """
    Fetch live positions and orders for all clients, print them, and return current positions.
    """
    current_positions = {}
    print("\n" + "=" * 80)
    print(f" Live Positions & Orders Update: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print("=" * 80)

    for client_id in client_details.keys():
        api = client_apis.get(client_id)
        if not api:
            print(f" No API instance found for client {client_id}. Skipping...")
            continue

        # Fetch data
        positions = get_client_live_positions(api, client_id)
        open_orders, rejected_orders = get_client_orders(api, client_id)

        # Store positions
        current_positions[client_id] = positions

        # Display results
        print(f"\n Client: {client_id}")
        print("-" * 40)

        # Display Positions
        print(" Positions:")
        if positions:
            for instr_id, qty in positions.items():
                symbol = master_global[master_global['instrumentID'] == str(instr_id)]['symbol1'].iloc[0] if instr_id in [int(x) for x in master_global['instrumentID']] else 'N/A'
                print(f"   - {symbol}: {qty}")
        else:
            print("    No positions available.")

        # Display Open Orders
        print(" Open Orders:")
        if open_orders:
            for order in open_orders:
                symbol = order.get('TradingSymbol', 'N/A')
                qty = order.get('OrderQuantity', 0)
                status = order.get('OrderStatus', 'N/A')
                print(f"   - {symbol}: Qty {qty}, Status: {status}")
        else:
            print("    No open orders.")

    print("\n Live data retrieval complete!")
    return current_positions

# --------------------------------------------------------
# Main Excel  Trading Input function (with position adjustment)
# --------------------------------------------------------
def excel_to_trading_input(input_path, current_positions):
    """
    1) Read user-order Excel
    2) Standardize 'NSE Symbol' and 'ISIN', then MERGE once against master_merge
       so that we know which rows match a valid instrument in master_instruments.xlsx.
    3) Iterate over valid rows to build stock_details, buy_details, sell_details,
       adjusting quantities based on current_positions.
    """
    # --- Load user orders into a DataFrame ---
    df = pd.read_excel(input_path, sheet_name='Sheet1', dtype=str)
    df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)

    # Standardize the key columns (uppercase & strip)
    df['NSE Symbol'] = df['NSE Symbol'].str.strip().str.upper()
    df['ISIN']       = df.get('ISIN', pd.Series()).fillna('').astype(str).str.strip().str.upper()

    # Add temporary columns for a vectorized merge
    df['merge_sym']  = df['NSE Symbol']
    df['merge_isin'] = df['ISIN']

    # --- Perform a single left-merge against master_merge ---
    merged = df.merge(
        master_merge,
        how='left',
        left_on=['merge_sym', 'merge_isin'],
        right_on=['symbol1', 'ISIN'],
        suffixes=('', '_master')
    )

    # Drop invalid rows where 'instrumentID' is NaN
    invalid_mask = merged['instrumentID'].isna()
    if invalid_mask.any():
        invalid_rows = merged.loc[invalid_mask, ['NSE Symbol', 'ISIN']]
        for _, row in invalid_rows.iterrows():
            print(f"WARNING: Input pair NSE Symbol='{row['NSE Symbol']}' + ISIN='{row['ISIN']}' not found in master_instruments.xlsx")
        merged = merged[~invalid_mask].copy()

    # --- Build the output dicts ---
    stock_details       = {}
    client_buy_details  = {}
    client_sell_details = {}

    def process_buy_sell(sub_df, is_buy: bool):
        """
        sub_df: subset of merged rows where either 'QTY to Buy' or 'QTY to Sell' is not null.
        is_buy: True => process 'QTY to Buy'; False => process 'QTY to Sell'
        """
        target = client_buy_details if is_buy else client_sell_details
        qty_col = 'QTY to Buy' if is_buy else 'QTY to Sell'

        for _, row in sub_df.iterrows():
            # Parse instrumentID, client ID, quantity, and SOV
            try:
                exch_id = int(row['instrumentID'])
                client  = row['USER ID'].strip()
                qty     = int(str(row[qty_col]).replace(',', '').strip())
                sov     = int(str(row['SOV']).replace(',', '').strip())
            except Exception as e:
                print(f"Skipping row due to parse error: {e}")
                continue

            # Get current position for this client and instrument
            P = current_positions.get(client, {}).get(exch_id, 0)

            # Adjust quantity based on current position
            if is_buy:
                if P >= 0:
                    adjusted_qty = max(qty - P, 0)
                else:
                    adjusted_qty = qty + abs(P)
            else:  # sell
                if P <= 0:
                    adjusted_qty = max(qty - abs(P), 0)
                else:
                    adjusted_qty = qty + P

            # Record stock_details if this is the first time we see this exch_id
            if exch_id not in stock_details:
                tick_size = float(row['ticker']) if row.get('ticker') not in [None, '', 'nan'] else 0.05
                stock_details[exch_id] = {
                    'name': row['NSE Symbol'],
                    'exchange_instrument_id': exch_id,
                    'tick_size': tick_size,
                    'high_string': row.get('HighString', ''),
                    'low_string': row.get('LowString', '')
                }

            # Record the client order with adjusted quantity
            if client not in target:
                target[client] = {
                    'single_order_value': sov,
                    f'total_quantity_to_{"buy" if is_buy else "sell"}': {}
                }
            key = f'total_quantity_to_{"buy" if is_buy else "sell"}'
            prev_qty = target[client][key].get(exch_id, 0)
            target[client][key][exch_id] = prev_qty + adjusted_qty

    # Process buy-side rows
    buys = merged[
        merged['QTY to Buy'].notna() &
        (merged['QTY to Buy'].astype(str).str.strip() != '')
    ]
    process_buy_sell(buys, is_buy=True)

    # Process sell-side rows
    sells = merged[
        merged['QTY to Sell'].notna() &
        (merged['QTY to Sell'].astype(str).str.strip() != '')
    ]
    process_buy_sell(sells, is_buy=False)

    # --- Filter out unused stock_details ---
    used_instruments = set()
    for details in client_buy_details.values():
        used_instruments |= set(details['total_quantity_to_buy'])
    for details in client_sell_details.values():
        used_instruments |= set(details['total_quantity_to_sell'])

    stock_details = {
        i: info
        for i, info in stock_details.items()
        if i in used_instruments
    }

    return stock_details, client_buy_details, client_sell_details

# --------------------------------------------------------
# Run as a script
# --------------------------------------------------------
if __name__ == "__main__":
    # Fetch current positions (fixing the typo: use client_apis instead of client_sell_details)
    current_positions = fetch_all_clients_data(client_apis)
    # Process Excel file with position adjustments
    stock_details, client_buy_details, client_sell_details = excel_to_trading_input("Sample Excel Inputalg.xlsx", current_positions)
    print("\nInstrument Details:", stock_details)
    print("Buy Orders:", client_buy_details)
    print("Sell Orders:", client_sell_details)
```
